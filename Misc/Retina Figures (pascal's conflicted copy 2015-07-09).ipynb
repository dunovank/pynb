# 2x plots with matplotlib and the IPython Notebook

This illustrates some basic steps to enabling 2x PNG matplotlib plots in the IPython notebook
for use with your fancy new double-density display.

```python
%pylab inline
```

Let's make a simple PNG plot:

```python
from matplotlib import pyplot as plt
import numpy as np

plt.plot(np.random.rand(100))
fig = plt.gcf()
```

Now double the DPI, so we are making 2x plots:

```python
import matplotlib
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
```

```python
plt.plot(np.random.rand(100))
fig2x = gcf()
```

But this PNG is now just twice as big and twice as ugly.
This is because the DPI of PNGs is generally completely ignored.

We need to do some tinkering to disable the raw PNG display of figures,
and replace it with simple HTML that scales for 2x display.

First, disable existing PNG formatters:

```python
import matplotlib

ip = get_ipython()

png_formatter = ip.display_formatter.formatters['image/png']
png_formatter.type_printers.pop(matplotlib.figure.Figure, None)
```

Now define our png2x function that displays figures as 2x PNGs via HTML:

```python
from base64 import encodestring

import matplotlib

from IPython.display import display, HTML
from IPython.core.pylabtools import print_figure

def png2x(fig):
    """render figure to 2x PNG via HTML"""
    x,y = matplotlib.rcParams['figure.figsize']
    dpi = matplotlib.rcParams['savefig.dpi']
    x2x = int(x * dpi / 2)
    y2x = int(y * dpi / 2)
    png = print_figure(fig, 'png')
    png64 = encodestring(png).decode('ascii')
    return u"<img src='data:image/png;base64,%s' width=%i height=%i/>" % (png64, x2x, y2x)
    
```

And let's test it out:

```python
display(HTML(png2x(fig2x)))
```

Hopefully that's about right.

Now let's register it as the default formatter:

```python
html_formatter = ip.display_formatter.formatters['text/html']
html_formatter.for_type(matplotlib.figure.Figure, png2x)
```

```python
display(fig2x)
```

The next step shouldn't be necessary, but until IPython is patched, this is a workaround.

What we are doing is disabling the post-execute function that automatically publishes all figures as PNG,
and replaces it with our own that publishes all figures with the plain display function.

```python
from matplotlib._pylab_helpers import Gcf

from IPython.zmq.pylab.backend_inline import flush_figures
ip._post_execute.pop(flush_figures, None)

def new_flush_figures():
    try:
        for figure_manager in Gcf.get_all_fig_managers():
            display(figure_manager.canvas.figure)
    finally:
        matplotlib.pyplot.close('all')

ip.register_post_execute(new_flush_figures)
```

Now we can just start plotting with 2x figures:

```python
plot(rand(100))
```
