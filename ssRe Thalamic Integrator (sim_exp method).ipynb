```python
conditions=['75', '250']
ssd_dict=dict()

for i, dfout in enumerate(df_ssd):
    
    conds_dict=dict()
    
    for df in dfout.keys():
        dfout[df].dropna(inplace=True)
        dfout[df].index=np.arange(len(dfout[df]))
        
        minlist=[]; maxlist=[]

        for trace in dfout[df][:]:
            csum_min=pd.Series(trace).cumsum().min()
            csum_max=pd.Series(trace).cumsum().max()
            minlist.append(csum_min)
            maxlist.append(csum_max)
        
        conds_dict[df]=(np.mean(minlist), np.mean(maxlist))
        
    ssd_dict[conditions[i]]=conds_dict
```
