```python
ssd_strings=[str(int(s*1000)) for s in ssdlist]

go_cor_conds=collections.OrderedDict()
stop_cor_conds=collections.OrderedDict()
go_err_conds=collections.OrderedDict()
stop_err_conds=collections.OrderedDict()

for i, dfout in enumerate(df_ssd):
    
    go_cor_conds[ssd_strings[i]]=dfout.ix[(dfout['choice']=='go')&(dfout['acc']==1), 'thalamus'].dropna()
    stop_cor_conds[ssd_strings[i]]=dfout.ix[(dfout['choice']=='stop')&(dfout['acc']==1), 'thalamus'].dropna()
    go_err_conds[ssd_strings[i]]=dfout.ix[(dfout['choice']=='go')&(dfout['acc']==0), 'thalamus'].dropna()
    stop_err_conds[ssd_strings[i]]=dfout.ix[(dfout['choice']=='stop')&(dfout['acc']==0), 'thalamus'].dropna()
    
gcond_cor_means={}; gcond_err_means={}
scond_cor_means={}; scond_err_means={}
    
for s in ssd_strings:

    gcond_cor_means[s]=np.mean([pd.Series(gtrace).cumsum().max() for gtrace in go_cor_conds[s][:]])
    gcond_err_means[s]=np.mean([pd.Series(gtrace).cumsum().max() for gtrace in go_err_conds[s][:]])
    scond_cor_means[s]=np.mean([pd.Series(strace).cumsum().max() for strace in stop_cor_conds[s][:]])
    scond_err_means[s]=np.mean([pd.Series(strace).cumsum().max() for strace in stop_err_conds[s][:]])

conditions={'CorrectGo':gcond_cor_means, 'SS_Stop': scond_cor_means, 
            'SS_Go':gcond_err_means, 'NoGoErrors': scond_err_means}
```
