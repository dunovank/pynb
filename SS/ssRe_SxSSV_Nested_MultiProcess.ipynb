# Functions to Optimize SSV for Nested Model

### Get Individual Sx Params for Simulating

```python
def get_sx_params(sxdf):
    
    params=sxdf['mean']; params.index=sxdf['param']
    a=params['a']*.1
    z=a*params['z']
    v=params['v(bsl)']*.1
    
    sp={'ssTer':0.0, 'ssTer_var':0.0, 'pGo':0.0}
    gp={'a':a, 'v':v, 'z':z, 'Ter':params['t'], 'eta':params['sv'], 'st':0.0, 'sz':0.0}
    
    return [gp, sp]
```

### Simulate & Residuals F(x)

```python
ss=reload(ss); qsim=reload(qsim); simfx=reload(simfx); utils=reload(utils); ft=reload(ft)

def ssvMinFunc(p, gp, sp, ydata, ntrials=1000):
    
    ssdlist=np.arange(.20, .45, .05)
    simdf_list=[]
    sp['mu_ss'] =  -p['mu_ss'].value
    
    for ssd in ssdlist:
        
        sp['ssd']=ssd
        out=ss.set_model(gParams=gp, sParams=sp, mfx=simfx.sim_radd, ntrials=ntrials, timebound=.650, 
                         depHyper=True, visual=False, task='ssRe', return_all_beh=True, condition_str='bsl') 
        simdf_list.append(out)
    
    simdf=pd.concat(simdf_list)
    ymodel=simdf.groupby('ssd').mean()['acc'].values
    
    residuals=ymodel-ydata
    #residuals = scp.sqrt(residuals ** 2 / ysigma ** 2)
    
    return residuals
```

### Gradient-Descent Simplex Optimization of SSV 

```python

def fitsx_ssv(output, sx, data, params, ssvinit, ntrials=5000, maxiter=500):
        
    sxdata=data[data['subj_idx']==sx]
    ydata=sxdata.groupby('ssd').mean()['acc'].values[:-1]
    #ysigma=sxdata.groupby('ssd')['acc'].std().values[:-1]

    gp, sp = get_sx_params(sxdf)
    p=Parameters()
    p.add('mu_ss', value=ssvinit, min=0.1, max=2.5)

    out = Minimizer(ssvMinFunc, p, fcn_args=(gp,sp,ydata,ntrials), 
                    kws={"full_output":True, "disp":True, "retall":True}) 
    out.prepare_fit()
    out.fmin(maxfun=maxiter, ftol=1.e-4, xtol=1.e-4)

    fits=OrderedDict([("sx",sx),('ssv',-p['mu_ss'].value), ('ssd',np.arange(.20, .45, .05)), 
                      ("residuals",out.residual),("ydata",ydata),('yhat',ydata+out.residual),
                      ("chisq",out.chisqr),("redchi",out.redchi)])
    
    #store subject fit summary in output queue
    output.put(fits)
```

# Commands to Multithread Subject-Level Fits

```python
# Define an output queue
output = mp.Queue()

# Setup a list of processes that we want to run
# target: function to multithread
# args: arguments for function
processes=[mp.Process(target=fitsx_ssv, args=(output, sx, data, sxdf, ssvinits[sx], 5000, 500)) for sx, sxdf in rePar.groupby('subj_idx')]
```

```python
# Run processes
for p in processes:
    p.start()
```

```python
# Exit the completed processes
for p in processes:
    p.join()
```
