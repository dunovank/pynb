```python
def integrator_stats(thalamus_traces, condition_list, fx=np.mean):
    
    cond_dict=collections.OrderedDict()
    
    for c in condition_list:
        
        c_int=[pd.Series(trace).cumsum().max() for trace in thalamus_traces[c][:]]
        
        cond_dict[c]=fx(c_int)
    
    return cond_dict
```

```python
pgo_index=[int(p*100) for p in pGo]

go_conds=collections.OrderedDict()
stop_conds=collections.OrderedDict()

for i, dfout in enumerate(dfv):

    go_conds[pgo_index[i]]=dfout.ix[dfout['choice']=='go', 'thalamus'].dropna()
    stop_conds[pgo_index[i]]=dfout.ix[dfout['choice']=='stop', 'thalamus'].dropna()

    
    
traces_dict={'go':go_conds, 'stop': stop_conds}
cmeans=collections.OrderedDict()
csems=collections.OrderedDict()

for ttype in traces_dict.keys():
    
    cmeans[ttype]=integrator_stats(traces_dict[ttype], pgo_index, fx=np.mean)
    csems[ttype]=integrator_stats(traces_dict[ttype], pgo_index, fx=sem)
```
