```python
pstop_lo_data = 0.0825056838578
pstop_hi_data = 0.720628415301
```

```python
qsim=reload(qsim); simfx=reload(simfx); ss=reload(ss); psy=reload(psy)
ssvlist = -1*np.arange(0.5,2.5,.25)
sim_pro = qsim.sim_ssv_range(ssvlist=ssvlist, task='ssPro', mfx=simfx.sim_radd, ntrials=500)
```

```python
LoSims=sim_pro[sim_pro['condition']=='Lo']
HiSims=sim_pro[sim_pro['condition']=='Hi']
pstop_lo_sims=np.array([ss.anl(lo_df)['pstop'] for ssv, lo_df in LoSims.groupby('ssv')])
pstop_hi_sims=np.array([ss.anl(hi_df)['pstop'] for ssv, hi_df in HiSims.groupby('ssv')])
```

```python
#f, ax = plt.subplots(1, figsize=(5, 4))
f = plt.figure(figsize=(8, 6)) 
sns.set_context('poster')
#f.subplots_adjust(top=0.95, wspace=0.12, left=0.19, right=0.98, bottom=0.15)	
ax = f.add_subplot(111)
sns.set_style('white')
#sns.set_context('poster')

x=ssvlist*-1
ax.plot(x, pstop_lo_sims,marker='o', 'Purple')
ax.plot(x, pstop_hi_sims, marker='o', 'LimeGreen')

lo_data=plt.hlines(y=pstop_lo_data, xmin=x[0], xmax=x[-1], lw=3, label="Low PGo Data", color='#441B5F')
hi_data=plt.hlines(y=pstop_hi_data, xmin=x[0], xmax=x[-1], lw=3, label="Hi PGo Data", color='#195E19')

ax.set_ylim=(pstop_lo_sims.min()-.1, pstop_lo_data+.1)
axx=plt.setp(ax, ylabel='P(Stop)', xlabel='|SS Drift|', xticks=x, xticklabels=x, xlim=(x[0]-.01, x[-1]+.01))
sns.despine(bottom=False, left=False, right=True, top=True)
```
