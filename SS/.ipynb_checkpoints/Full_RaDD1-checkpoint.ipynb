```python
def get_sx_params(sxdf):
    
    params=sxdf['mean']; params.index=sxdf['param']
    a=params['a']*.1
    z=a*params['z']
    v=params['v(bsl)']*.1
    
    sp={'ssTer':0.0, 'ssTer_var':0.0, 'pGo':0.0}
    gp={'a':a, 'v':v, 'z':z, 'Ter':params['t'], 'eta':params['sv'], 'st':0.0, 'sz':0.0}
    
    return [gp, sp]
```
