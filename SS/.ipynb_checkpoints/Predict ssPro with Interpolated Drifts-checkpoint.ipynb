```python
from __future__ import division
import os
from collections import OrderedDict
import time
import seaborn as sns
import pandas as pd
from radd import qsim, psy, utils, ft, simfx, ss
from radd.simfx import *

from lmfit import minimize, Parameters, Parameter, report_fit, Minimizer, fit_report
import scipy.interpolate as intpol
```

```python
pth=utils.find_path()

rePar=pd.read_csv(pth+"CoAx/SS/HDDM/Reactive/vbias_full/vBP_SxStats.csv")
proPar=pd.read_csv(pth+"CoAx/SS/HDDM/Proactive/vfull_sx/vfull_HiLo_550_SxStats.csv")
#data=pd.read_csv(pth+"CoAx/SS/Reactive/Re_AllData.csv")
data=pd.read_csv(pth+"CoAx/SS/Proactive/Pro_SxData_tb550.csv")
```

```python
sim_rts=[]; sim_curves=[]

for sx, sxdf in df.groupby('subj_idx'):
    sim_rts.append(np.array([ss.anl(sxdf[sxdf['pGo']==pg])['go_rt_all'] for pg in pGo_list]))
    sim_curves.append(np.array([ss.anl(sxdf[sxdf['pGo']==pg])['pstop'] for pg in pGo_list]))
```
