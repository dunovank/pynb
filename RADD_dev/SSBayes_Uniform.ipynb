```python
from radd.tools.analyze import kde_fit_quantiles
def breakdown_vector(vector, fitparams):
    
    ssd, nssd, nss, nss_per, ssd_idx = fitparams['ssd_info']
    nquant = fitparams['quantiles'].size
    
    presp = vector[1:1+nssd]
    quant = vector[-nquant*2:]
    quant_cor = quant[:nquant]
    quant_err = quant[-nquant:]
    qc_kde = kde_fit_quantiles([quant_cor], bw=.01)
    qe_kde = kde_fit_quantiles([quant_err], bw=.01)
    return presp, qc_kde, qe_kde
```
