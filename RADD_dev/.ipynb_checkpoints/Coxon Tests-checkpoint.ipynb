```python
def go_stop_accuracy(lines=[],  yerr=[], ax=None, linestyles=None, colors=None, labels=None, x=None):

    if ax is None:
        f, ax = plt.subplots(1, figsize=(5.5, 6))
    if colors is None:
        colors = ['Green', 'Red']
    if labels is None:
        labels = [''] * len(lines)
    if linestyles is None:
        linestyles = ['-'] * len(lines)
    if x is None:
        x = np.arange(len(lines[0]), dtype='float')
    for i, yi in enumerate(lines):
        divider = make_axes_locatable(ax)
        axx = divider.append_axes("top", size=.5, pad=0.01, sharex=ax)
        # Plot the results
        if yerr != []:
            axx.errorbar(x, yi, yerr=yerr[i], color=colors[i], lw=0, marker='o', elinewidth=2, ecolor=colors[i])
            axx.plot(xsim, pxp, linestyle=linestyles[i], lw=2., color=colors[i], label=labels[i])
        else:
            axx.plot(x, yi, marker='o', linestyle=linestyles[i], lw=2.1, color=colors[i], label=labels[i])

    plt.setp(axx, xticks=x, ylim=(-.01, 1.05), yticks=np.arange(0,1.2,.2))#[0,  1])
    axx.set_xticklabels(['Go', 'Stop'])
    axx.set_ylabel('Accuracy')
    axx.legend(loc=0)
    plt.tight_layout()
    sns.despine()
    return ax
```
