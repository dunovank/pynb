```python
f,ax=plt.subplots(figsize=(6,5))

xxticks=np.arange(0,1.2,.2)

v_chivec=vboot.groupby('pGo').mean()['chi'].values
t_chivec=tboot.groupby('pGo').mean()['chi'].values
v_chivec_err=vboot.groupby('pGo').sem()['chi'].values*1.96
t_chivec_err=tboot.groupby('pGo').sem()['chi'].values*1.96

#ax.errorbar(xxticks, v_chivec, yerr=v_chivec_err, color=reds[5], ecolor='k', label='Drift Bias', lw=3, elinewidth=2)
#ax.errorbar(xxticks, t_chivec, yerr=t_chivec_err, color=purples[5], ecolor='k', label='Onset Bias', lw=3, elinewidth=2)

ax.plot(xxticks, v_chivec, color=reds[5], label='Drift Bias', lw=3)
ax.plot(xxticks, t_chivec, color=purples[5], label='Onset Bias', lw=3)

vchi_table=vboot.pivot_table('chi', 'n', 'pGo')
tchi_table=tboot.pivot_table('chi', 'n', 'pGo')

for n in xrange(len(vchi_table)):
    plt.plot(xxticks, vchi_table.iloc[n, :], color=reds[5], lw=0, marker='o', ms=4, alpha=.1)
    plt.plot(xxticks, tchi_table.iloc[n, :], color=purples[5], lw=0, marker='o', ms=4, alpha=.1)
    

ax.set_xticks(xxticks)
ax.set_xticklabels(xxticks, fontsize=18)
plt.setp(ax.get_yticklabels(), fontsize=15)
xxticklabels=ax.set_xticklabels([str(xx) for xx in np.arange(0,1.2,.2)])
#yylabel=ax.set_ylabel('$ \chi^2 (+/- 95\% CI) $', fontsize=22)
yylabel=ax.set_ylabel('$ \chi^2$', fontsize=22)
xxlabel=ax.set_xlabel('P(Go)', fontsize=22)
ax.set_xlim(-.1, 1.1)
ax.set_ylim(-.001, .025)
ax.legend(loc=0, fontsize=20)
plt.tight_layout()
sns.despine()

f.savefig(outpath+"ProBoot_AllPGo_MeanChi_Onset_vs_Drift.png", format="png", dpi=300)
f.savefig(outpath+"ProBoot_AllPGo_MeanChi_Onset_vs_Drift.svg", format="svg", rasterized=True, dpi=300)
```
