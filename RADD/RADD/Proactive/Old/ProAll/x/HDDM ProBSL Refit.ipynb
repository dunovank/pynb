```python
prodf=pd.read_csv(pth+"DataFiles/Pro_Data.csv", index_col=0)

prodf_cut=prodf[prodf.rt<=.545]
sns.set_style('ticks')
f,axes=plt.subplots(2,3, figsize=(12,8)); i=0
cp=sns.color_palette(n_colors=6)
for pg, pgdf in prodf_cut.groupby('pGo'):
    ax=f.axes[i]
    pgdf.rt.hist(label="PGo"+str(pg*100)[:-2], bins=20, ax=f.axes[i], color=cp[i])
    
    pgdf.rt.hist(label="RT="+str(pgdf.rt.mean())[:5], bins=20, ax=f.axes[i], color=cp[i])
    sns.despine()
    ax.legend(loc=2, fontsize=13)
    plt.setp(ax, xlim=(.3,.55))
    i+=1   
plt.tight_layout()
f.subplots_adjust(top=.92)
f.suptitle("RT <= 545 msec", fontsize=22)
f.savefig("ssPro_RT_Cut <= 545 msec.png", dpi=600)
```
