```python
import radd
from radd import utils, vis, pbold, final
import scipy.stats as stats
csv=pd.read_csv
from scipy.stats import ttest_rel

labels=['Baseline', 'Difficult']
colors=['#2d2d2d','#6a6a6a']
out="/Users/kyle/Desktop/"
```

```python
pro_all=final.all_data('pro')
bsl=pro_all.query('Cond=="bsl"')
pnl=pro_all.query('Cond=="pnl"')
```

```python
bsl5451=utils.get_5451df(bsl)
pnl5451=utils.get_5451df(pnl)

bsl_pstop_idx=1-bsl5451.pivot_table('response', 'idx', 'pGo')
pnl_pstop_idx=1-pnl5451.pivot_table('response', 'idx', 'pGo')

bsl_pstop=bsl_pstop_idx.mean()
pnl_pstop=pnl_pstop_idx.mean()

bsl_pstop_err=bsl_pstop_idx.sem()*1.96
pnl_pstop_err=pnl_pstop_idx.sem()*1.96
```

```python
vis.scurves([bsl_pstop.values, pnl_pstop.values], yerr=[bsl_pstop_err.values, pnl_pstop_err.values], colors=colors, labels=labels, task='pro', sxdata=[bsl_pstop_idx, pnl_pstop_idx])

plt.savefig(out+"pro_sc.png", format='png', dpi=600)
plt.savefig(out+"pro_sc.svg", rasterized=True, dpi=600)
```

```python
psebsl=vis.scurves(bsl_pstop_idx.values, pse_only=True, task='Pro')
psepnl=vis.scurves(pnl_pstop_idx.values, pse_only=True, task='Pro')
```

```python
print np.mean(psebsl)
print np.mean(psepnl)

print stats.sem(psebsl)*1.96
print stats.sem(psepnl)*1.96
```

```python
means_pse=[np.mean(psebsl), np.mean(psepnl)]
pse_err=[stats.sem(psebsl)*1.96, stats.sem(psepnl)*1.96]
sxdata=[psebsl, psepnl]
ax=vis.plot_re_bar(means_pse, yerr=pse_err, sxdata=sxdata, yylim=[.3, .80], ylabel='PSE', colors=colors)
plt.savefig(out+'pre_pse.png', dpi=600, format='png')
plt.savefig(out+'pre_pse.svg', dpi=600, format='svg')
```

```python
print "bsl/pnl PSE means: ", means_pse
print "t:%.4f\np:%.4f"% ttest_rel(sxdata[0], sxdata[1])
```
