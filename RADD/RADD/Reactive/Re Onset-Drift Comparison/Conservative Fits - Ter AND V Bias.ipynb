```python
def visualize_model_params(paramdf=pd.DataFrame, save=False, savestr="model_id", diffplot=False, ax=None, dist=False, bw=.08):
    
    x=np.arange(len(paramdf.columns))
    xnames=list(paramdf.columns)
    
    if diffplot:
        xnames=['$\Delta$'+i for i in xnames]

    mu=paramdf.mean()
    err=paramdf.sem()*1.96
    sns.set_style('ticks', rc=sp['style'])
    c=sns.color_palette('hls', n_colors=len(x))
    
    if ax is None and dist:
        f, (ax, ax2) = plt.subplots(1,2, sharey=True, squeeze=True)
        sns.violinplot(tv_diff[['t', 'v']], ax=ax2, positions=x, bw=bw, color=c, inner='box')
        err=np.zeros_like(err)

    elif ax is None and not dist:    
        f, ax = plt.subplots(1)
    
    #c=sns.color_palette('hls', n_colors=len(x))
    
    for xx in x:
        ax.bar(xx, mu[xx], yerr=err[xx], color=c[xx], ecolor='k', align='center')
        if dist:
            ax.plot(np.array([xx]*len(paramdf)), paramdf.iloc[:, xx], marker='o', ms=3.5, lw=0, color='k', alpha=.3)
    
    sns.despine()
    
    if dist:
        ylim=(paramdf.min().min()-.01, paramdf.max().max()+.01)
    else:
        ylim=(mu.values.min()-.02, mu.values.max()+.02)
        
    for a in f.axes:
        a.set_xlabel("Parameter %s"%'$\\theta$', fontsize=22)
        plt.setp(a, xticks=x, xticklabels=xnames)
        plt.setp(a.get_yticklabels(), fontsize=16)
        plt.setp(a.get_xticklabels(), fontsize=20)
        plt.tight_layout()
    
    ax.hlines(y=0, xmin=ax.get_xlim()[0], xmax=ax.get_xlim()[1], alpha=.4, linestyle='--')
    ax.set_ylabel('$\\theta$'+" Units", fontsize=22)
    ax.set_ylim(ylim)

    if save:
        f.savefig(savestr+"_params.png", format='png', dpi=600)
        f.savefig(savestr+"_params.svg", rasterized=True, dpi=600)
    
    return f
```
