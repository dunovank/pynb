```python
def visualize_model_params(paramdf=pd.DataFrame, save=False, savestr="model_id", diffplot=False, ax=None):
    
    x=np.arange(len(paramdf.columns))
    xnames=list(paramdf.columns)
    
    if diffplot:
        xnames=['$\Delta$'+i for i in xnames]
    #xnames[-1]='$\Delta$'+'t'
    mu=paramdf.mean()
    err=paramdf.sem()*1.96
    
    sns.set_style('ticks', rc=sp['style'])
    
    if ax is None:
        f, ax = plt.subplots(1, figsize=(7.5, 5))
    
    c=sns.color_palette('hls', n_colors=len(x))
    
    for xx in x:
        ax.bar(xx, mu[xx], yerr=err[xx], color=c[xx], ecolor='k', align='center')
    sns.despine()

    ax.set_xlabel("Parameter %s"%'$\\theta$', fontsize=22)
    ax.set_ylabel('$\\theta$'+" Units", fontsize=22)
    plt.setp(ax, xticks=x, xticklabels=xnames, xlim=(-.8,x[-1]+.8), ylim=(mu.values.min()-.2, mu.values.max()+.2))

    plt.setp(ax.get_yticklabels(), fontsize=16)
    plt.setp(ax.get_xticklabels(), fontsize=20)
    plt.hlines(y=0, xmin=-2, xmax=x[-1]+3, alpha=.4, linestyle='--')
    plt.tight_layout()

    if save:
        f.savefig(savestr+"_params.png", format='png', dpi=600)
        f.savefig(savestr+"_params.svg", rasterized=True, dpi=600)
    
    return ax
```

```python
def visualize_model_params(paramdf=pd.DataFrame, save=False, savestr="model_id", diffplot=False, ax=None):
    
    x=np.arange(len(paramdf.columns))
    xnames=list(paramdf.columns)
    
    if diffplot:
        xnames=['$\Delta$'+i for i in xnames]
    #xnames[-1]='$\Delta$'+'t'
    mu=paramdf.mean()
    err=paramdf.sem()*1.96
    
    sns.set_style('ticks', rc=sp['style'])
    
    if ax is None:
        f, ax = plt.subplots(1, figsize=(7.5, 5))
    
    c=sns.color_palette('hls', n_colors=len(x))
    
    for xx in x:
        ax.bar(xx, mu[xx], yerr=err[xx], color=c[xx], ecolor='k', align='center')
    sns.despine()

    ax.set_xlabel("Parameter %s"%'$\\theta$', fontsize=22)
    ax.set_ylabel('$\\theta$'+" Units", fontsize=22)
    plt.setp(ax, xticks=x, xticklabels=xnames, xlim=(-.8,x[-1]+.8), ylim=(mu.values.min()-.2, mu.values.max()+.2))

    plt.setp(ax.get_yticklabels(), fontsize=16)
    plt.setp(ax.get_xticklabels(), fontsize=20)
    plt.hlines(y=0, xmin=-2, xmax=x[-1]+3, alpha=.4, linestyle='--')
    plt.tight_layout()

    if save:
        f.savefig(savestr+"_params.png", format='png', dpi=600)
        f.savefig(savestr+"_params.svg", rasterized=True, dpi=600)
    
    return ax







def compare_model_effects(y, yerr=[0,0], save=False, savestr='vt'):
    
    x=np.arange(len(y))+1
    
    xnames=['$\Delta$'+'v','$\Delta$'+'t']
    yy=np.sort(y); yyerr=np.sort(yerr)
    yyticks=[yy[0]-yyerr[0]-.01, 0.00, yy[-1]+yyerr[-1]+.01]
    yystr=[str(yyy)[:5] for yyy in yyticks]
    c=sns.color_palette('hls', n_colors=6)
    colors=[c[4], c[1]]

    f, ax = plt.subplots(1, figsize=(4.5,5.5))

    ax.bar(x, y, yerr=yerr, align='center', color=colors, ecolor='k')

    plt.setp(ax, xticks=x, yticks=yyticks, xlim=(0.5,2.5), ylim=(yyticks[0], yyticks[-1]))
    ax.set_yticklabels(yystr, fontsize=22)
    ax.set_xticklabels(xnames, fontsize=26)
    ax.set_xlabel("Penalty Bias", fontsize=26)
    ax.set_ylabel('$\\theta$'+" Units", fontsize=26)
    plt.hlines(y=0, xmin=0, xmax=3, alpha=.4, linestyle='--')

    sns.despine()
    plt.tight_layout()
    
    if save:
        outpath=utils.find_path()+"FinalFigures_New/"
        f.savefig(outpath+"Figure3A_ModelComparison_"+savestr+".png", format='png', dpi=600)
        f.savefig(outpath+"Figure3A_ModelComparison_"+savestr+".svg", rasterized=True, dpi=600)
    
    return ax
```
