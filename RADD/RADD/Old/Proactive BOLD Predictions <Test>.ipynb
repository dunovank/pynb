# Drift-Bias Proactive BOLD Predictions

```python
from __future__ import division
from radd import utils, predict, fitpro, fitre, fitfx, pbold
from pandas import read_csv as csv
import scipy as scp

pth=utils.find_path()

fitpath=pth+"FitFiles/"
parampath=pth+"ParamFiles/"
outpath=pth+"FinalFigures/"

#prosc_data=np.array([ 0.95280437,  0.87844952,  0.66402309,  0.40945818,  0.071913, 0.0066407 ])
pro_sc = lambda df: 1-df.groupby('pGo').mean()['response'].values
pro_rt = lambda df: df.query('acc==1 & choice=="go" & rt < .5451').groupby('pGo').mean()['rt'].values
pgo = np.arange(0, 1.25, .25)
```

```python
bxv=csv(parampath+"rwr_pro_fitinfo.csv")
pro=csv(parampath+"rwr_pro_opt_theta.csv")
prodict=dict(pro.iloc[0:, :].values)

params=['ssv', 'z', 'a', 't', 'ssd']
theta={p:prodict[p] for p in params}

theta_fresh=theta.copy()
```

# Drift-Bias vs. Onset-Bias

### simulate drift-bias bold data

```python
theta=theta_fresh.copy()
vbias = {'v': bxv.groupby('pGo').mean()['v'].values}
boldv = pbold.probold_bias(theta, bias=vbias, timebound=.560, ntrials=150, ssv_decay=True, return_all=True)
```

### simulate onset-bias bold data

```python
theta=theta_fresh.copy()
tlist = np.linspace(.2, .35, len(pgo))[::-1]
tbias = {'t': tlist}; theta['v']=1.6484
boldt = pbold.probold_bias(theta, bias=tbias, timebound=.560, ntrials=150, ssv_decay=True, return_all=True)
```

### SSV ON during decay epoch

```python
axv=pbold.plot_bold_manuscript(boldv, title='Drift Bias', save=False, bias="drift")
#axt=pbold.plot_bold_manuscript(boldt, title='Onset Bias', save=False, bias='onset')
```
