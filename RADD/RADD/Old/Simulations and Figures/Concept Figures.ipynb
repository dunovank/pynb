```python
from __future__ import division
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#globals
a=.37; z=.5*a; Ter=.347; eta=.14; st=.0001; sz=.0001; s2=.01; xpo=[12, 12.29]; pSSD=.450;
mu_ss=-2.17; ssTer=.099; ssRe_TB=.653; ssPro_TB=.6; nt=3000; sTB=.00001; ssTer_var=.0001
```

```python
noise = np.random.normal(0,.15,2000)
direct_start=np.zeros(50)
direct = np.linspace(0,1,1000)
direct_end = np.ones(950)


indirect_start = np.zeros(200)
indirect = 0 - np.linspace(0,.35,1200)
indirect_end = np.ones(600)*-.35


ss_noise = np.random.normal(0, .15, 2000)
ss_start = np.zeros(1000)
ss = 0 - np.linspace(0,1,200)
ss_end = np.ones(800)*ss[-1]
```

```python
direct_signal = np.concatenate((direct_start, direct, direct_end)).flatten()
direct_signal = direct_signal + noise

indirect_signal = np.concatenate((indirect_start, indirect, indirect_end)).flatten()
indirect_signal = indirect_signal + noise

ss_signal = np.concatenate((ss_start, ss, ss_end)).flatten()
ss_signal = ss_signal + ss_noise
```

```python
sns.set(style="white", context='poster')
f=plt.figure(figsize=(5, 10))
axgo = plt.subplot2grid((3, 2), (0, 0), rowspan=1, colspan=2)
axng = plt.subplot2grid((3, 2), (1, 0), rowspan=1, colspan=2)
axss = plt.subplot2grid((3, 2), (2, 0), rowspan=1, colspan=2)


axgo.plot(np.linspace(0,1,2000), direct_signal, lw=.5, color='LimeGreen')
axng.plot(np.linspace(0,1,2000), indirect_signal, lw=.5, color='RoyalBlue')
axss.plot(np.linspace(0,1,2000), ss_signal, lw=.5, color='Red')
sns.despine(bottom=True)

for ax in f.axes:
    ax.set_ylim(-1.5, 1.5)
    ax.hlines(y=0.0, xmin=0, xmax=1, lw=5, color='k',linestyle='--', alpha=.5)
    for label in ax.xaxis.get_ticklabels():
        label.set_visible(False)
```
