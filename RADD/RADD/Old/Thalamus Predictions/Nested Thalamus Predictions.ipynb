```python
def integrator_stats(thalamus_traces, condition_list, fx=np.mean):
    
    cond_dict=collections.OrderedDict()
    
    for c in condition_list:
        
        c_int=[pd.Series(trace).cumsum().max() for trace in thalamus_traces[c][:]]
        
        cond_dict[c]=fx(c_int)
    
    return cond_dict
```
