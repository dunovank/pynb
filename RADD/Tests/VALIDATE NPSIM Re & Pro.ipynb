# SIM RE (IDX PARAMS)

```python
sim_rebsl_rt, sim_rebsl_sc = {},{}
sim_repnl_rt, sim_repnl_sc = {},{}
delays=np.arange(0.200, 0.450, .05)  

for sx in vbsl.index.unique():
    
    sxtheta_bsl=vbsl.loc[sx, :].to_dict()
    sxtheta_pnl=vpnl.loc[sx, :].to_dict()
    
    for theta in [sxtheta_bsl, sxtheta_pnl]:
        theta['pGo']=.5
    
    sxbsl_sacc, sxpnl_sacc, sxbsl_rt, sxpnl_rt = [], [], [], []
    
    for ssd in delays:
        sxtheta_bsl['ssd']=ssd
        rebsl_dvg, rebsl_dvs = npsim.radd(sxtheta_bsl, ntrials=5000, timebound=.650)
        bsl_sacc, bsl_rt =  npsim.analyze_reactive_trials(rebsl_dvg, rebsl_dvs, sxtheta_bsl)
        sxbsl_sacc.append(bsl_sacc)
        sxbsl_rt.append(bsl_rt)
        
        sxtheta_pnl['ssd']=ssd
        repnl_dvg, repnl_dvs = npsim.radd(sxtheta_pnl, ntrials=5000, timebound=.650)
        pnl_sacc, pnl_rt =  npsim.analyze_reactive_trials(repnl_dvg, repnl_dvs, sxtheta_pnl)
        sxpnl_sacc.append(pnl_sacc)
        sxpnl_rt.append(pnl_rt)
        
    sim_rebsl_sc[sx]=sxbsl_sacc; sim_rebsl_rt[sx]=sxbsl_rt
    sim_repnl_sc[sx]=sxpnl_sacc; sim_repnl_rt[sx]=sxpnl_rt

sim_rebsl_sc=pd.DataFrame.from_dict(sim_rebsl_sc, orient='index').mean().values
sim_repnl_sc=pd.DataFrame.from_dict(sim_repnl_sc, orient='index').mean().values
sim_rebsl_rt=pd.DataFrame.from_dict(sim_rebsl_rt, orient='index').mean().values.mean()
sim_repnl_rt=pd.DataFrame.from_dict(sim_repnl_rt, orient='index').mean().values.mean()
```
