```python
def __update_params__(sim, p):
      """ update Pg (probability of DVg +dx) and Tg (num go process timepoints)
      for go process and get get dynamic bias signal if 'x' model
      """
      Pg = 0.5*(1 + p['v']*sim.dx/sim.si)
      Ps = 0.5*(1 + p['ssv']*sim.dx/sim.si)
      Tg = np.ceil((sim.tb-p['tr'])/sim.dt).astype(int)
      Ts = np.ceil((sim.tb-p['sso'])/sim.dt).astype(int)
      nt = np.max(hs([Tg, Ts]))
      t = np.cumsum([sim.dt]*nt)
      xtb = sim.temporal_dynamics(p, t)
      return Pg, Tg, Ps, Ts, xtb
```
