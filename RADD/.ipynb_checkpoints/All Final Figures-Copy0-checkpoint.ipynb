# Define Summary Fx

```python
def summarize_bx_sc(df, stat='mean'):
    
    if stat=='mean':
        val=df.iloc[:,:-1].mean().values
    elif stat=='std':
        val=df.iloc[:,:-1].std().values*1.96
    elif stat=='sem':
        val=df.iloc[:,:-1].sem().values*1.96
    return val

def summarize_bx_rt(df, stat='mean'):
    if stat=='mean':
        val=df.iloc[:,-1].mean()
    elif stat=='std':
        val=df.iloc[:,-1].std()*1.96
    elif stat=='sem':
        val=df.iloc[:,-1].sem()*1.96
        
    return val


def get_reactive_stops(df, errors='sem'):
    
    ys=pd.DataFrame(columns=df.ssdlist.unique()[:-1], index=df.idx.unique())
    for ix, ixdf in df.groupby('idx'):
        ys.loc[ix, :]=ixdf.ydata.values[:-1]
    
    avg=df.groupby('ssdlist').mean()
    
    if errors=='sem':
        err=df.groupby('ssdlist').sem()
    elif errors=='std':
        err=df.groupby('ssdlist').std()
        
    yd=avg['ydata'][:-1]
    yh=avg['yhat'][:-1]
    yderr=err['ydata'][:-1]*1.96
    yherr=err['yhat'][:-1]*1.96
    
    return yd, yh, yderr, yherr, ys.values


def get_reactive_rts(df, errors='sem'):
    
    ys=(df[df['ssdlist']=='rt'].loc[:,'ydata']/10)*1000
    ysh=(df[df['ssdlist']=='rt'].loc[:,'yhat']/10)*1000
    ys.index=df.idx.unique()
    
    avg=df.groupby('ssdlist').mean()*1000
    if errors=='sem':
        err=df.groupby('ssdlist').sem()*1000
    elif errors=='std':
        err=df.groupby('ssdlist').std()*1000
        
    
    rtd=avg['ydata'][-1]/10
    rth=avg['yhat'][-1]/10
    
    rtd_err=err['ydata'][-1]/10*1.96
    rth_err=err['yhat'][-1]/10*1.96
    
    return rtd, rth, rtd_err, rth_err, ys, ysh
```
