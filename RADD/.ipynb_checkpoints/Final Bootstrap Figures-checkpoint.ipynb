# Define Summary Fx

```python

def summarize_bx_sc(df, stat='mean'):
    
    if stat=='mean':
        val=df.iloc[:,:-1].mean().values
    elif stat=='ci':
        val=df.iloc[:,:-1].std().values*1.96
    return val

def summarize_bx_rt(df, stat='mean'):
    if stat=='mean':
        val=df.iloc[:,-1].mean()
    elif stat=='ci':
        val=df.iloc[:,-1].std()*1.96
    return val


def get_reactive_stops(df):
    

    ys=pd.DataFrame(columns=df.ssdlist.unique()[:-1], index=df.idx.unique())
    for ix, ixdf in df.groupby('idx'):
        ys.loc[ix, :]=ixdf.ydata.values[:-1]
    
    avg=df.groupby('ssdlist').mean()
    sem=df.groupby('ssdlist').sem()

    yd=avg['ydata'][:-1]
    yh=avg['yhat'][:-1]
    yderr=sem['ydata'][:-1]*1.96
    yherr=sem['yhat'][:-1]*1.96
    
    return yd, yh, yderr, yherr, ys


def get_reactive_rts(df):
    
    ys=(df[df['ssdlist']=='rt'].loc[:,'ydata']/10)*1000
    ysh=(df[df['ssdlist']=='rt'].loc[:,'yhat']/10)*1000
    ys.index=df.idx.unique()
    
    avg=df.groupby('ssdlist').mean()*1000
    sems=df.groupby('ssdlist').sem()*1000
    
    rtd=avg['ydata'][-1]/10
    rth=avg['yhat'][-1]/10
    
    rtd_err=sems['ydata'][-1]/10*1.96
    rth_err=sems['yhat'][-1]/10*1.96
    
    return rtd, rth, rtd_err, rth_err, ys, ysh
```

```python
redat_sc=redat_df.mean().values[:-1]
redat_rt=redat_df.mean().values[-1]/10
redat_sc_ci=redat_df.sem().values[:-1]*1.96
redat_rt=redat_df.mean().values[-1]/10
redat_rt_ci=redat_df.sem().values[-1]*1.96/10

restop=rebsl[rebsl['trial_type']=='stop']
sxsacc=utils.makePivot(restop)

#only plot rts for subjects we eventually fit (58 out of 61)
sxrts=(simbsl[simbsl['ssdlist']=='rt'].loc[:,'ydata']/10)*1000
sxrts.index=simbsl.idx.unique()

emp_b, sim_b, emp_berr, sim_berr, bslsx_ps=get_reactive_stops(simbsl)
#emp_p, sim_p, emp_perr, sim_perr, pnlsx_ps=get_reactive_stops(simpnl)
```
