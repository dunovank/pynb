# Drift-Bias Proactive BOLD Predictions

```python
from __future__ import division
from radd import utils, predict, fitpro, fitre, fitfx, pbold
from pandas import read_csv as csv
import scipy as scp

pth=utils.find_path()

fitpath=pth+"FitFiles/"
parampath=pth+"ParamFiles/"
outpath=pth+"FinalFigures/"

#prosc_data=np.array([ 0.95280437,  0.87844952,  0.66402309,  0.40945818,  0.071913, 0.0066407 ])
pro_sc = lambda df: 1-df.groupby('pGo').mean()['response'].values
pro_rt = lambda df: df.query('acc==1 & choice=="go" & rt < .5451').groupby('pGo').mean()['rt'].values
pgo = np.arange(0, 1.25, .25)
```

```python
bxv=csv(parampath+"rwr_pro_fitinfo.csv")
pro=csv(parampath+"rwr_pro_opt_theta.csv")
prodict=dict(pro.iloc[0:, :].values)

params=['ssv', 'z', 'a', 't', 'ssd']
theta={p:prodict[p] for p in params}

theta_fresh=theta.copy()
```

# Drift-Bias vs. Onset-Bias

### SSV ON during decay epoch

```python
theta=theta_fresh.copy()
vbias = {'v': bxv.groupby('pGo').mean()['v'].values}
boldv = pbold.probold_bias(theta, bias=vbias, timebound=.560, ntrials=2000, ssv_decay=False, return_all=True)

axv=pbold.plot_bold_manuscript(boldv, title='Drift Bias (ssv OFF)', save=True, bias="drift_ssvOFF")
axv_all=pbold.plot_bold_all(boldv, title='Drift Bias (ssvOFF)', bias='drift_ssvOFF', save=True)

#----------------------------------------------------------------------------------------------------------------#

theta=theta_fresh.copy()
tbias = {'t': np.linspace(.1, .45, len(pgo))[::-1]}
theta['v']=1.6484
boldt = pbold.probold_bias(theta, bias=tbias, timebound=.560, ntrials=2000, ssv_decay=False, return_all=True)

axt=pbold.plot_bold_manuscript(boldt, title='Onset Bias (ssv OFF)', save=True, bias='TR_ssvOFF')
axt_all=pbold.plot_bold_all(boldt, title='Onset Bias (ssvOFF)', bias='TR_ssvOFF', save=True)
```

### SSV OFF during decay epoch

```python
theta=theta_fresh.copy()
vbias = {'v': bxv.groupby('pGo').mean()['v'].values}
boldv = pbold.probold_bias(theta, bias=vbias, timebound=.560, ntrials=2000, ssv_decay=True, return_all=True)

axv2=pbold.plot_bold_manuscript(boldv, title='Drift Bias (ssv ON)', save=True, bias="drift_ssvON")
axv2_all=pbold.plot_bold_all(boldv, title='Drift Bias (ssvON)',bias='drift_ssvON', save=True)
#----------------------------------------------------------------------------------------------------------------#

theta=theta_fresh.copy()
tbias = {'t': np.linspace(.1, .45, len(pgo))[::-1]}
theta['v']=1.6484
boldt = pbold.probold_bias(theta, bias=tbias, timebound=.560, ntrials=2000, ssv_decay=True, return_all=True)

axt2=pbold.plot_bold_manuscript(boldt, title='Onset Bias (ssv ON)', save=True, bias='TR_ssvON')
axt2_all=pbold.plot_bold_all(boldt, title='Onset Bias (ssvON)',bias='TR_ssvON', save=True)
```
