# Model Comparison: Stop Curves

```python
dflist=[bsldata, bslfits, ipbfits, ipafits]
sc_means=np.array([df['sc'].mean().values for df in dflist])
sc_cis=np.array([df['sc'].sem().values for df in dflist])*1.96

labels=['Baseline', 'Dep. Pools', 'IP Equal', 'IP Matched']
colors=["#2d2d2d",'#1abd80', "#E74C3C", '#FFBC39']
vis.scurves(sc_means, yerr=sc_cis, colors=colors, labels=labels, task='Re', linestyles=['-']*4)

plt.savefig(out+"radd_ip_scfits.png", dpi=600, format='png')
plt.savefig(out+"radd_ip_scfits.svg", dpi=600, format='svg', rasterized=True)
```

# Model Comparison: RTs
