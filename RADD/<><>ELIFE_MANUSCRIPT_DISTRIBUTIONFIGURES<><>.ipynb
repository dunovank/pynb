```python
def get_rts(sim, DV, p):

      nss = m.simulator.nss; prob = m.simulator.prob
      ssd = m.simulator.ssd; tb = m.simulator.tb
      nc = m.simulator.ncond; nssd=m.simulator.nssd
      
      DVg, DVs = DV
      
      gdec = m.simulator.resp_up(DVg, p['a'])
      gort = m.simulator.RT(p['tr'], gdec)
      sdec = m.simulator.resp_lo(DVs)
      ssrt = m.simulator.RT(ssd, sdec)
      
      errs = gort[:, :5000][:,None]*np.ones_like(ssrt)
      
      ert = np.nanmean(np.where(errs<ssrt, 1, np.nan)*errs, axis=1)
      crt = np.where(gort<tb, 1, np.nan)*gort
      return crt, ert
```
