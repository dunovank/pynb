```python
sns.()
```

```python
for i, yhat in enumerate(yhats):
      sns.set_context('notebook', font_scale=1.6)
      f, axes = plt.subplots(1,3,figsize=(11, 4))
      fitsc, fitcq, fiteq = vis.unpack_yvector(yhat)
      
      data_kde = data_kde
      axes_x, pp = vis.plot_data_dists(data=data_kde, cdf=False, axes=axes, data_type='interpolated')
      ax1, ax2, ax3 = axes_x
      fit_cq, fit_eq = [analyze.kde_fit_quantiles(q, bw=.1) for q in [fitcq, fiteq]]

      sns.kdeplot(fit_cq, color=colors[i], cumulative=False, bw=.1, ax=ax1, linewidth=3, alpha=.60, shade=False)      sns.kdeplot(fit_eq, color=colors[i], cumulative=False, bw=.001, ax=ax2, linewidth=3, alpha=.60, shade=False)
      vis.scurves([sc, fitsc], kind='radd', linestyles=['-','--'], ax=ax3, colors=[colors[i]]*2)
      for ax in [ax1, ax2]:
            if ax.is_last_col():
                  continue
            ax.set_xlim(.4, .65)
            if ax.is_first_col():
                  ax.set_ylabel('P(RT)')
            ax.set_xlabel('RT (ms)')
            ax.set_xticklabels([int(xx) for xx in ax.get_xticks()*1000])

            ax.set_ylim(0,10)
      ax1.text(.41, 9, 'Correct', fontsize=19)
      ax2.text(.41, 9, 'Error', fontsize=19)
      plt.tight_layout()
      plt.subplots_adjust(hspace=.5)
      #plt.savefig(savestrings[i]+'.png', dpi=900)
      #plt.savefig(savestrings[i]+'.svg', rasterized=True)
```
