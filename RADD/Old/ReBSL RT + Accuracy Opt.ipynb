```python
opt=reload(opt)

def init_ssv(issv, y, theta, step_eval=25, crit=.05, bounce=.05):

    searching=True
    neval=0
    orig_step=step_eval
    issv=-abs(issv)
    
    residuals=opt.raddMinFunc(issv, theta, ntrials=500, ydata=y, fitting=False)
    
    while searching:

        adj=abs(issv)*bounce
        
        if np.mean(residuals)>0:
            #make more positive/weaker
            issv=issv+adj
        else:
            #make more negative/stronger
            issv=issv-adj

        residuals=opt.raddMinFunc(issv, theta, ntrials=500, ydata=y, fitting=False)
        sse=np.sum([i**2 for i in residuals])
        
        if neval>step_eval:
            crit+=.025
            step_eval+=step_eval

            print "current sse : ", sse
            print "increasing error tolerance to : ", crit
            
            if step_eval>3*orig_step:
                bounce=bounce+bounce*.5
                print "increasing step size..."
            
        neval+=1
        
        if sse<crit:
            adj_ssv=issv
            searching = False
    
    return adj_ssv
            
```
