```python
def plot_correlations(x, y, ax, colors=[]):
    
    if colors==[]:
        i=np.random.randint(50)
        colors=sns.color_palette('hls', n_colors=50, desat=True)
    
    pcor=np.corrcoef(x,y)[0][1]
    m,b=np.polyfit(x,y,1)

    corr_str="%s: r=%s" % (parameter, str(pcor)[:5])
    ax.scatter(x, y, color=colors[i])
    ax.plot(x, m*x+b,color=colors[i], label=corr_str)
    ax.legend(loc=2, fontsize=14)

    sns.despine()

    return ax
```

```python
def set_labels(f, xlabel, ylabel):
    
    for ax in f.axes:
        
        if ax.is_first_col():
            plt.setp(ax, ylabel='RaDD')
        if ax.is_last_row():
            plt.setp(ax, xlabel='HDDM')
    
    plt.tight_layout()
    
    return f
```
