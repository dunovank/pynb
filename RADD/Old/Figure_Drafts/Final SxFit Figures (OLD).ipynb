```python
def find_path():

    if os.path.isdir("/Users/kyle"):
        pth="/Users/kyle/Dropbox/"

    elif os.path.isdir("/home/kyle"):
        pth="/home/kyle/Dropbox/"

    return pth

def simConditions(theta, ntrials=1500, tb=.650, task='ssRe'):
    
    theta['pGo']=.5
    theta['ssv']=-abs(theta['ssv'])
    simdf_list=[]; ssdlist=np.arange(.2,.45,.05)
    for ssd in ssdlist:     
        theta['ssd']=ssd
        
        out=fitre.simulate(theta, ntrials=ntrials, depHyper=True, visual=True, return_all=True) 
        vis.plot_traces(df=out, theta=theta, pGo=.5, ssd=ssd, timebound=0.653, task='ssRe')
        simdf_list.append(out[['rt', 'choice', 'response', 'acc', 'trial_type', 'ssd', 'pGo']])
        
    df=pd.concat(simdf_list)
    return df

def get_reactive_stops(df):
    
    ys=pd.DataFrame(columns=df.ssdlist.unique()[:-1], index=df.idx.unique())
    for ix, ixdf in df.groupby('idx'):
        ys.loc[ix, :]=ixdf.ydata.values[:-1]
    
    avg=df.groupby('ssdlist').mean()
    sem=df.groupby('ssdlist').sem()

    yd=avg['ydata'][:-1]
    yh=avg['yhat'][:-1]
    yderr=sem['ydata'][:-1]*1.96
    yherr=sem['yhat'][:-1]*1.96
    
    return yd, yh, yderr, yherr, ys


def get_reactive_rts(df):
    
    ys=(df[df['ssdlist']=='rt'].loc[:,'ydata']/10)*1000
    ysh=(df[df['ssdlist']=='rt'].loc[:,'yhat']/10)*1000
    ys.index=df.idx.unique()
    
    avg=df.groupby('ssdlist').mean()*1000
    sems=df.groupby('ssdlist').sem()*1000
    
    rtd=avg['ydata'][-1]/10
    rth=avg['yhat'][-1]/10
    
    rtd_err=sems['ydata'][-1]/10*1.96
    rth_err=sems['yhat'][-1]/10*1.96
    
    return rtd, rth, rtd_err, rth_err, ys, ysh

def makePivot(df, cols='ssd', index=None, values='acc', func=np.mean):
    pvot=pd.pivot_table(df, index=index, columns=cols, values=values, aggfunc=func)
    return pvot

def pgo(df):
    return makePivot(df, cols='pGo', index='idx', values='response')

def pstop(df):
    return 1-makePivot(df, cols='pGo', index='idx', values='response')

def pro_rt(df):
    godf=df[df['choice']=='go']
    return makePivot(godf, cols='pGo', index='idx', values='rt')

def re_rt(df):
    corr=df[(df['choice']=='go')&(df['acc']==1)]
    return makePivot(corr, cols='ssd', index='idx', values='rt')

def re_acc(df):
    stops=df[df['trial_type']=='stop']
    return makePivot(stops, index='idx', cols='ssd', values='acc')

styl=utils.style_params()
sns.set_style('ticks', rc=styl['style'])
sns.set_context('paper', rc=styl['context'])
sns.set_palette(styl['colors'])
```
