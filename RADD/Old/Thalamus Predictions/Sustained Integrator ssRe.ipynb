```python
from __future__ import division
from radd import ss, psy, simfx
from radd.simfx import sim_radd, sim_ss, sustained_integrator, integrator
from scipy.stats import sem
import seaborn as sns
import pandas as pd
import os
import collections
from itertools import cycle
```

```python
def integrator_stats(thalamus_traces, condition_list, fx=np.mean):
    
    cond_dict=collections.OrderedDict()
    
    for c in condition_list:
        
        c_int=[pd.Series(trace).cumsum().max() for trace in thalamus_traces[c][:]]
        
        cond_dict[c]=fx(c_int)
    
    return cond_dict
```

```python
ssd_index=[int(s*1000) for s in ssdlist]

go_cor_conds=collections.OrderedDict()
stop_cor_conds=collections.OrderedDict()
go_err_conds=collections.OrderedDict()
stop_err_conds=collections.OrderedDict()

for i, dfout in enumerate(df_ssd):
    
    go_cor_conds[ssd_index[i]]=dfout.ix[(dfout['choice']=='go')&(dfout['acc']==1), 'thalamus'].dropna()
    stop_cor_conds[ssd_index[i]]=dfout.ix[(dfout['choice']=='stop')&(dfout['acc']==1), 'thalamus'].dropna()
    go_err_conds[ssd_index[i]]=dfout.ix[(dfout['choice']=='go')&(dfout['acc']==0), 'thalamus'].dropna()
    stop_err_conds[ssd_index[i]]=dfout.ix[(dfout['choice']=='stop')&(dfout['acc']==0), 'thalamus'].dropna()

    
traces_dict={'SS_Stop':stop_cor_conds, 'NoGoErrors':stop_err_conds, 'CorrectGo':go_cor_conds, 'SS_Go':go_err_conds}
cmeans=collections.OrderedDict()
csems=collections.OrderedDict()

for ttype in traces_dict.keys():
    
    cmeans[ttype]=integrator_stats(traces_dict[ttype], ssd_index, fx=np.mean)
    csems[ttype]=integrator_stats(traces_dict[ttype], ssd_index, fx=sem)
```

```python
means=pd.DataFrame(cmeans).fillna(0)
sems=pd.DataFrame(csems).fillna(0)
```
