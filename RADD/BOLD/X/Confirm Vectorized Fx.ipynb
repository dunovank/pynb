```python
from radd import utils, vis, vectorize_fitfx
from pandas import read_csv as csv
pth=utils.find_path()
fitpath=pth+"FitFiles/"
parampath=pth+"ParamFiles/"
outpath=pth+"FinalFigures/"

#prosc_data=np.array([ 0.95280437,  0.87844952,  0.66402309,  0.40945818,  0.071913, 0.0066407 ])
pro_sc = lambda df: 1-df.groupby('pGo').mean()['response'].values
pro_rt = lambda df: df.query('acc==1 & choice=="go" & rt < .5451').groupby('pGo').mean()['rt'].values
bxv=csv(parampath+"rwr_pro_fitinfo.csv")
vvals=bxv.groupby('pGo').mean()['v'].values

pro=csv(parampath+"rwr_pro_opt_theta.csv", header=None)
params=['ssv', 'z', 'a', 't', 'ssd']
theta={p:pro[pro[0]==p].values[0][1] for p in params}
```

```python
prosc_df=pd.read_csv(pth+"DataFiles/pro_sc.csv", index_col=0)
prort_df=pd.read_csv(pth+"DataFiles/pro_rt.csv", index_col=0)
```

```python
sc_data=np.array([ 0.95280437,  0.87844952,  0.66402309,  0.40945818,  0.071913  ,  0.0066407 ])
sc_data_ci=np.array([ 0.0134097 ,  0.02133956,  0.03718259,  0.04255428,  0.02868479,  0.00444822])
rt_data=np.array([ 0.53095071,  0.53440574,  0.53247535,  0.52135209,  0.51488091])
rt_data_ci=np.array([ 0.00681328,  0.00272061,  0.0027018,  0.00333285,  0.00395515])
```

```python
pgo=np.arange(0,1.2,.2)
alldf=pd.DataFrame(columns=['rt', 'pstop'], index=pgo)

for i, pg in enumerate(pgo):
    
    theta['v']=vvals[i]

    go, stop = vectorize_fitfx.vectorized_proactive_radd(theta, model='radd', ntrials=5000, timebound=0.560)
    
    rt, pstop = vectorize_fitfx.analyze_proactive_vectors(go, stop, theta)
    
    alldf.loc[pg, 'rt']=rt
    alldf.loc[pg, 'pstop']=pstop
```

```python
os.chdir(outpath)

sns.set_context(rc={'lines.markeredgewidth': 0.1})

sc_radd=alldf['pstop'].values.astype('float')
sc_radd_ci=np.array([ 0.01259545,  0.02612138,  0.03428849,  0.03652605,  0.02016176,
        0.0131357 ])

vis.scurves([sc_data, sc_radd], yerr=[sc_data_ci, sc_radd_ci], colors=['#363636', '#fc2654'], labels=['Data', 'RADD'], task='ssPro', sxdata=prosc_df.values)

plt.savefig("proboot_vectorized_SC_95CIonSD.png", dpi=600)
plt.savefig("proboot_vectorized_SC_95CIonSD.svg", rasterized=True, dpi=600)
```

```python
os.chdir(outpath)
rt_radd=alldf['rt'].values[1:]
rt_radd_ci=np.array([ 0.00197428,  0.00142838,  0.00103825,  0.00150899,  0.00184657])

vis.plot_pro_rt(y=[rt_data, rt_radd], yerr=[rt_data_ci, rt_radd_ci], colors=['#363636','#fc2654'], labels=['Data', 'RADD'], sxdata=prort_df)

plt.savefig("proboot_vectorized_RT_95CIonSD.png", format='png', dpi=600)
plt.savefig("proboot_vectorized_RT_95CIonSD.svg", rasterized=True, dpi=600)
```
