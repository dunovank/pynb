```python
def store_recost(indxi, label, params=None, i=None, yhat=None):
      # get predictions and store optimized parameter set
      #params['indx']=indxi; params[self.cond]=label
      gqpi = np.hstack([indxi,label,'go', yhat[:5]*.1, yhat[5]])
      eqpi = np.hstack([indxi, label, 'stop', yhat[6:11]*.1, yhat[11]])
      pstopi = np.hstack([indxi, label, yhat[12:]])
      #popti = pd.Series({k:params[k] for k in self.popt.columns})

      # fill df: [cond] [go/ssgo], [cor/err rt quantiles], [prob corr/err response]
      gqp_fits.loc[i,:] = gqpi
      eqp_fits.loc[i,:] = eqpi
      pstop_fits.loc[i,:] = pstopi
      i+=1
```
