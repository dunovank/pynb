```python
rts = [df.query('acc==1 & trial_type=="go"').rt.values*1000 for df in [rebsl, repnl]]

def reactive_rt_distributions(rts, style='kde', save=False):
    
    sns.set_context(context='paper', font_scale=2)

    labels=['BSL', 'PNL']
    allcolors= sns.color_palette('Greens', n_colors=10)[3:]
    recolors = [allcolors[1], allcolors[-3]]
    xxlim = [400,680]

    if style == 'violin':
        f, ax = plt.subplots(1, figsize=(6,4))
    else:
        f, axes = plt.subplots(1, 2, figsize=(10,3))
        plt.subplots_adjust(wspace=.8)


    for i, rt in enumerate(rts):

        c = recolors[i]
        low, high = xxlim[0], xxlim[1]
        lbl=labels[i]

        if style == 'kde':
            axx = axes.flatten()[i]
            ax = sns.kdeplot(rt,shade=True, clip=(low,high), alpha=.8, bw=6, label=lbl, lw=2, color=c, ax=axx)

        elif style == 'violin':
            if i==0:
                pass
            else:
                ax = sns.violinplot(rts, positions=[i, i+1], names=labels, lw=2, bw=.1, widths=.4)
                ax.set_ylabel('RT (ms)')
                ax.set_yticks([350,750])
                ax.set_yticklabels([str(yy) for yy in [350,750]])
                ax.set_ylim(350,750)

        elif style == 'dist':
            axx = axes.flatten()[i]
            ax = sns.distplot(rt, kde=False, bins=50, hist_kws={'alpha':.9, "color":c, 'label':lbl}, ax=axx)

        axx.legend(loc=0)

    if not style == 'violin':

        for i, ax in enumerate(axes.flatten()):

            ax.set_xlim(low,high)
            yylim = [0, ax.get_ylim()[1]+(ax.get_ylim()[1]*.15)]
            ax.set_ylim(yylim)
            ax.set_xticks(xxlim)
            ax.set_xticklabels(xxlim)
            ax.set_xlabel('RT (ms)')

            ax.set_yticks([])
            ax.set_yticklabels([])

            if style == 'kde' and i==0:
                ax.set_ylabel('Probability Mass')
            elif style == 'dist' and i==0:
                ax.set_ylabel('Probability')
                ax.set_yticks(yylim)
                ax.set_yticklabels([str(yy) for yy in yylim])

            plt.tight_layout()
            sns.despine()

        if style=='kde' and save:
            pth="/home/kyle/Dropbox/"
            plt.savefig(pth+"reactive_rt_kde_subplots.svg", format='svg', rasterized=True, dpi=600)
            plt.savefig(pth+"reactive_rt_kde_subplots.png", dpi=600)
        elif style=='dist' and save:
            plt.savefig(pth+"reactive_rt_dist_subplots.svg", format='svg', rasterized=True, dpi=600)
            plt.savefig(pth+"reactive_rt_dist_subplots.png", dpi=600)
    
    elif style == 'violin' and save:
        plt.savefig(pth+"reactive_rt_violin_subplots.svg", format='svg', rasterized=True, dpi=600)
        plt.savefig(pth+"reactive_rt_violin_subplots.png", dpi=600)
    plt.subplots_adjust(wspace=.5)
```
