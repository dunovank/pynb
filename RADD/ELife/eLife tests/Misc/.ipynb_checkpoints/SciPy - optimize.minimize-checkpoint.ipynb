```python
from scipy.optimize import minimize
```

```python
# residual function
fucks = lambda p,x,y: np.sum((y-(-x**p[0]+p[1]))**2)

# true function
fx = lambda p, x: -x**p[0]+p[1]
```

```python
# DV
x=np.arange(50)
#parameters: p
m=1.5; b=60; p=[m, b]
# true function
fx = lambda p, x: -x**p[0]+p[1]

# draw "true" line with true function
yline = fx(p, x)
# create some data
data = yline + norm.rvs(0, 90, len(x))

# provide initial guess
inits = [1.2, 60]
# optimize starting from inits
fits = minimize(fucks, inits, args=(x, data), method='Nelder-Mead')
# draw predicted line with fitted params
yhat_line = fx(fits.x, x)


#plot 
f, ax = plt.subplots(1, figsize=(4,5))
sns.set_context(font_scale=1.6, context='paper')
ax.plot(x, yline, color=reds[2], lw=3, label="$fx(Fucks)=-%.fFucks+%.f$"%(m, b))
ax.scatter(x, data, color='k', s=10, label='Fucks')
ax.plot(x, yhat_line, color=blues[1], ls='--', lw=3, label="yhat")

#make it pretty and labeled
plt.legend(loc=0, fancybox=True, frameon=True); sns.despine()
plt.setp(ax, xlabel='Lifespan', ylabel='# Fucks Left to Give', xticklabels=[], xticks=[])
```
