```python
def simple_resim(theta, ssdlist=range(200, 450, 50), ntrials=2000, return_all=False, model='radd'):
    return RADD.run(theta, tb=.650, ntrials=ntrials, model=model)
```

```python
theta = pd.read_csv('CoAx/SS/Reactive/BSL/Boot/RADD/rwr_rebsl_popt_radd.csv', index_col=0).mean().to_dict()
theta['pGo']=.5
theta['ab']=.1*theta['a']
theta['ssd']=.35
```

```python
DVg, DVs = simple_resim(theta, model='abias')
```

```python
model='abias'
tb=.650

ngo=len(DVg)
nss=len(DVs)

#theta=utils.update_params(theta)
tr=theta['tt'];	a=theta['a']; ssd=theta['ssd']; z=theta['z']
```

```python
#get individual trial go and ss rt's
#gos_i = [tr + np.argmax(DVgn>=a)*.0005 if np.any(DVgn>=a) else 999 for i, DVgn in enumerate(DVg)]
theta=utils.update_params(theta)
tr=theta['tt'];	a=theta['a']; ssd=theta['ssd']

if model=='abias':
    ab=a+theta['ab']
    Tg=np.ceil((tb-tr)/.0005)
    Ts=np.ceil((tb-ssd)/.0005)
    delay = Tg - Ts
    
    #get individual trial go and ss rt's
    GoRT_err = [tr + np.argmax(vec>=ab)*.0005 if np.any(vec[vec[delay:]>=ab]) else 999 for i, vec in enumerate(DVg[:nss])]
    GoRT_cor = [tr + np.argmax(vec>=a)*.0005 if np.any(vec[vec[:delay]>=a]) else 999 for i, vec in enumerate(DVg[nss:])]
    GoRT=np.append(GoRT_err, GoRT_cor)

if model in ['radd', 'ipb','abias']:
        stops_i = [ssd + np.argmax(DVsn<=0)*.0005 if np.any(DVsn<=0) else 999 for i, DVsn in enumerate(DVs)]
else:
        stops_i = [ssd + np.argmax(DVsn>=a)*.0005 if np.any(DVsn>=a) else 999 for i, DVsn in enumerate(DVs)]

#calculate stop accuracy
stops = np.array([1 if GoRT[i]>si else 0 for i, si in enumerate(stops_i)])
gos = np.append(np.abs(1-stops), np.ones(ngo - nss))
sacc=np.mean(stops)
```

```python
def plot_traces(DVg, DVs, sim_theta, tau=.0005, tb=.5451, cg='Green', cr='Red'):
    
    nss=len(DVs)
    f,ax=plt.subplots(1,figsize=(8,5))
    tr=sim_theta['t']; a=sim_theta['a']; z=sim_theta['z']; ssd=sim_theta['ssd']

    for i, igo in enumerate(DVg[nss:]):        
        gtrace = igo[igo<=a]
        xg = tr+np.arange(len(gtrace))*.0005
        plt.plot(xg, gtrace, color=cg, linewidth=.5, alpha=.1)
    
    if 'ab' in sim_theta.keys():
        a=a+sim_theta['ab']
    
    for i, igo in enumerate(DVg[:nss]):
        gtrace = igo[igo<=a]
        strace = DVs[i][DVs[i]>=0]
        
        xg = tr+np.arange(len(gtrace))*.0005
        plt.plot(xg, gtrace, color=blues[3], linewidth=.5, alpha=.1)
        
        #if i<len(DVs):
        xs = ssd+np.arange(len(strace))*.0005
        #return xs
        plt.plot(xs, strace, color=cr, alpha=.1, linewidth=.5)
    xlim=(tr-(.25*tr), tb)    
    plt.setp(ax, xlim=xlim)
    ax.hlines(y=z, xmin=xlim[0], xmax=tb, linewidth=3, linestyle='--', color="#6C7A89")
    ax.hlines(y=a, xmin=ssd, xmax=tb, linewidth=3, linestyle='-', color="#e74c3c")
    sns.despine(top=False,bottom=False, right=True, left=False)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_xticks([])
    ax.set_yticks([])

    return ax
```
