```python
from __future__ import division
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from radd import utils

%matplotlib inline
%load_ext retina
%load_ext autoreload
%autoreload 2

pd.set_option('display.max_rows', 30)
pd.set_option('max_colwidth', 20)
pd.set_option('precision', 5)


from radd.misc.colors import style_params
sp = style_params()
rcdict={'text.color': '#222222', 'axes.labelcolor': '#222222', 'figure.facecolor': '#dddddd'}
sns.set_style('ticks', rc=rcdict)
colors=sp['colors']; reds=sp['reds']; greens=sp['greens']; blues=sp['blues']; purples=sp['purples']
yellows=sp['yellows']; grays=sp['grays']

if os.path.isdir("/Users/kyle"):
      pth="/Users/kyle/Dropbox/"

elif os.path.isdir("/home/kyle"):
      pth="/home/kyle/Dropbox/"

os.chdir(pth)

rpal = lambda nc: sns.blend_palette(['#e88379', '#c0392b'], n_colors=nc)
bpal = lambda nc: sns.blend_palette(['#81aedb', '#4168B7'], n_colors=nc)
gpal = lambda nc: sns.blend_palette(['#65b88f', '#16a085'], n_colors=nc)
ppal = lambda nc: sns.blend_palette(['#848bb6', '#674172'], n_colors=nc)

"""
                                                #REACTIVE
"""

redata = pd.read_csv(pth+'ELife/Data/re_data.csv', index_col=0)
rebsl = redata.query('Cond=="bsl"')
repnl = redata.query('Cond=="pnl"')

rebsl_ipa_popt = pd.read_csv(pth+'CoAx/SS/Reactive/BSL/Boot/IPA/rwr_rebsl_popt_ipa.csv', index_col=0).mean().to_dict()
rebsl_idx_popt = pd.read_csv(pth+"CoAx/SS/Reactive/BSL/Idx/RADD/bsl_allsx_popt_radd.csv", index_col=0)
repnl_idx_popt = pd.read_csv(pth+"CoAx/SS/Reactive/PNL/Idx/RADD/pnl_allsx_popt_radd.csv", index_col=0)
rebsl_boot_popt = pd.read_csv(pth+'CoAx/SS/Reactive/BSL/Boot/RADD/rwr_rebsl_popt_radd.csv', index_col=0).mean().to_dict()
repnl_boot_popt = pd.read_csv(pth+'CoAx/SS/Reactive/PNL/Boot/RADD/pnl_rwr_radd_popt_radd.csv', index_col=0).mean().to_dict()

redf_store = pd.Series({"redata": redata, "rebsl" : rebsl, "repnl" : repnl,
      "rebsl_idx_popt" : rebsl_idx_popt, "repnl_idx_popt" : repnl_idx_popt,
      "rebsl_boot_popt" : rebsl_boot_popt, "repnl_boot_popt" : repnl_boot_popt,
      "rebsl_ipa_popt":rebsl_ipa_popt})


"""
                                                #PROACTIVE
"""

pro = pd.read_csv(pth+'ELife/Data/pro_data.csv')
proimg = pd.read_csv(pth+'ELife/Data/pro_img_data.csv')
vpopt = pd.read_csv(pth+"CoAx/SS/Proactive/Boot/drift-bias/flat_opt_theta.csv", header=None, names=['mean'], index_col=0).to_dict()['mean']
tpopt = pd.read_csv(pth+"CoAx/SS/Proactive/Boot/onset-bias/flat_opt_theta.csv", header=None, names=['mean'], index_col=0).to_dict()['mean']

vboot_info = pd.read_csv(pth+'CoAx/SS/Proactive/Boot/drift-bias/rwr_pro_fitinfo.csv')
tboot_info = pd.read_csv(pth+'CoAx/SS/Proactive/Boot/onset-bias/rwr_pro_fitinfo.csv')

proinits_df = pd.read_csv(pth+"CoAx/SS/Proactive/Boot/pro_popt_df.csv",index_col='param')
v = proinits_df.query('model=="drift"')
vtheta, vopt = utils.get_proactive_params(v.to_dict()['popt'], dep='v')
t = proinits_df.query('model=="onset"')
ttheta, topt = utils.get_proactive_params(t.to_dict()['popt'], dep='tr')

prodf_store = pd.Series({"prodata" : pro, 'proimg': proimg, 'vtheta' : vtheta, 'ttheta':ttheta, "vboot_info" : vboot_info, "tboot_info" : tboot_info, 'topt':topt, "tpopt" : tpopt,  'vopt':vopt, "vpopt" : vpopt})
```

```python
dirpath = pth+'ELife/IRM_V_Avg/'
if not os.path.isdir(dirpath):
      os.mkdir(dirpath)
os.chdir(dirpath)

data = redf_store.redata
inits = dict(deepcopy(redf_store.rebsl_ipa_popt))
```

```python
m = build.Model(data=data, inits=inits, kind='radd', fit_on='average', depends_on={'v': 'Cond'})
```

```python
#m.optimize(ntrials=20000, ftol=1.e-4, xtol=1.e-4, maxfev=100000)
m.optimize(ntrials=20000, ftol=1.e-2, xtol=1.e-2, maxfev=100)
```

```python
yhat = m.simulate().reshape(2,16)
for i in range(m.ncond):
      vis.plot_fits(m.avg_y[i], yhat[i], plot_acc=True, save=True, savestr=m.labels[i]+'_sim')
```
