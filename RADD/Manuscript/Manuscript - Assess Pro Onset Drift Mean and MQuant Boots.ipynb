```python
def get_quantiles_error(df):
    quantp=[.1,.3,.5,.7,.9]; 
    godf=df.query('choice=="go"')
    ixlist, quant_errs=[], []
    for pg, pgdata in godf.groupby('pGo'):
        for sx, sxdata in pgdata.groupby('idx'):
            ixq=np.mean(mquantiles(sxdata.rt.values, prob=quantp))
            ixlist.append(ixq)
        quant_errs.append(sem(ixlist))
        
    return quant_errs
```

```python
def plot_chi_dist(chi, bins=15, labels=['Drift', 'Onset'], rug=False, norm=False, colors=['#1abd80', 'Purple'], hist=True, kde=False, hist_kws=None, kde_kws=None):

    f,ax=plt.subplots(1, figsize=(4.5, 4.5))
    if kde==True:
        kde_kws={'shade':True, 'alpha':.6}
        hist=False
    if hist:
        hist_kws={'alpha':.8}
    if labels==[]:
        labels=[None, None]
        
        
    for i, ichi in enumerate(chi):
        ax=sns.distplot(ichi, bins=bins, rug=rug, color=colors[i], label=labels[i], kde=kde, hist=hist, 
                        ax=ax, hist_kws=hist_kws, kde_kws=kde_kws, norm_hist=norm)
    if labels[0]:
        ax.legend(fontsize=20)

    ax.set_xlabel('$\chi^2$', fontsize=28)
    ylim=[0, int(ax.get_yticks()[-1])]
    xlim=[0, ax.get_xticks()[-1]]
    plt.setp(ax, xlim=xlim, xticks=xlim, ylim=ylim, yticks=ylim)
    
    if norm:
        ytl=[0,1]
        yl='Probability'
    else:
        ytl=ylim
        yl='Count'
    
    ax.set_ylabel(yl, fontsize=24)
    ax.set_yticklabels(ytl, fontsize=22)
    ax.set_xticklabels(xlim, fontsize=22)

    plt.tight_layout()
    
    sns.despine()
    
    return ax
```

```python
def plot_bar_lines(bars, lines, berr, lerr, labels=['Data', 'Drift', 'Onset'], colors=['#2d2d2d', '#1abd80', purples[5]], dist_list=[], plot_lines=False):
    
    f, ax=plt.subplots(1, figsize=(6,5))
    x=np.arange(20,120,20)
    #for i, m in enumerate(means):
    ax.bar(x, bars, yerr=berr, width=10, align='center', color=colors[0], error_kw=dict(elinewidth=2, capsize=0, ecolor='k'), label=labels[0], alpha=1)
    
    if plot_lines:
        ax.errorbar(x, lines[0], yerr=lerr, color=colors[1], lw=3, label=labels[1])
        ax.errorbar(x, lines[1], yerr=lerr, color=colors[2], lw=3, label=labels[2])
    
    dist1, dist2=dist_list[:]
    for n1 in dist1:
        ax.plot(x, n1, color=colors[1], lw=0, alpha=.3, markersize=5, marker='o')
    
    for n2 in dist2:
        ax.plot(x, n2, color=colors[2], lw=0, alpha=.3, markersize=5, marker='o')
    
    if not plot_lines:
        ax.plot(x, n1,color=colors[1], lw=0, alpha=1, markersize=5, marker='o', label=labels[1])
        ax.plot(x, n2, color=colors[2], lw=0, alpha=1, markersize=5, marker='o', label=labels[2])
    
    ax.legend(loc=0, fontsize=20)
    yylim=[.49, .56]
    plt.setp(ax, ylim=(int(yylim[0]*1000),int(yylim[1]*1000)), yticks=yylim, xticks=x)
    ax.set_yticklabels(yylim, fontsize=22)
    ax.set_xticklabels([str(int(xt)) for xt in x], fontsize=22)
    ax.set_ylabel('RT (ms)', fontsize=24)
    ax.set_xlim(10, 110)
    
    plt.tight_layout()
    sns.despine()
    
    return ax
```
