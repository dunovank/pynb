```python
import radd
from radd import build
from radd import vis
import numba as nb
from numba import jit
from radd.tools.utils import PBinJ
from time import sleep
%matplotlib inline
```

```python
# read in example dataframe 
data = radd.load_example_data()
# baseline (bsl) and caution (pnl) conditions in 'Cond' column
```

```python
# initiate dependent process model with dynamic gain (kind = 'xdpm')
# (change kind='dpm' for model with no dynamic gain signal)
# drift-rate (v) depends on Baseline and Caution levels of 'Cond' variable
m = build.Model(kind='xdpm', data=data, depends_on={'v':'Cond'}, fit_on='average')
```

```python
m.fitparams['y']=m.observed[0]
m.fitparams['wts']=m.cond_wts[0]
m.fitparams['nlevels']=2
m.__set_ssd_info__()
y = m.fitparams['y']
yhat = y.flatten() + np.random.random_sample(y.size)*.01
```
