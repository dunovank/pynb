# Let's define some variables

```python
# x is an 'int' (or integer in regular speak)
x = 1
print 'x is', type(x)
```

# What does __type 'int'__ mean?
---
* everything in python is an object
    * all variables 
    * all functions
    * all classes
    * by everything i mean everything
---
* x is an object, and its *type* is __int__
* we can use the built-in __type()__ function to find this out



```python
# y is the name of a 'list' object. The list contains two 'int' objects
# we could just as well as named our list "asdfasdfasdf" and it would still be a list of ints

y = [1, 2]
print "y is a %s of %s's" % (type(y), type(y[0]))
```

```python
# Now we'll make another list, called z, filled with three 'str' or string objects
z = ['one', 'two', 'three']
print "z is a %s of %s's" % (type(z), type(z[0]))
```

# Working with lists

```python
# acess the first element (python indexes from 0)
print z[0]

# acess the second element
print z[1]

# replace the last element
z[-1]=3000
print z[-1]

# make a new list from the combination of two others
l = y + z
print l
```

#Objects have methods

Methods basically describe the functionality of a particular object. 
For instance, you can add an element to an existing list by using its __append__ method. 

We can do this (below) by inserting our int object, __x__, at the end of our list object, __y__. 

However, we couldn't do this the other way around because __append__ is a special method that only
list objects can perform. Try __x.append(y)__ and you'll see what I mean.

```python
print 'before', y
y.append(x)
print 'after', y
```
