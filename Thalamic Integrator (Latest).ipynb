```python
conditions=['75', '250']
ssd_dict=dict()

for i, dfout in enumerate(df_ssd):
    
    conds_dict=dict()
    
    for df in dfout.keys():
        dfout[df].dropna(inplace=True)
        dfout[df].index=np.arange(len(dfout[df]))
        
        csumlist=[]

        for trace in dfout[df][:]:
            csum=pd.Series(trace).cumsum().max()
            csumlist.append(csum)
        
        conds_dict[df]=np.mean(csumlist)
        
    ssd_dict[conditions[i]]=conds_dict
```
