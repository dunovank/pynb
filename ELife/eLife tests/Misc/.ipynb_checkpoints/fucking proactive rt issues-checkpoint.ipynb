```python
def rtdist(data, group='pGo', filtr=.53, bins=20, dist='h', c=None, prob=np.arange(.1,.9,.1), xlim=(.41,.58)):
    
    data['pGo'] = data.pGo.astype(float)
    data['HML'] = data.HML.astype(int)
    df = data.query('response==1 & pGo>.1')
    numcond=len(df[group].unique())
    if dist in ['c', 'qp']:
        nax=1
    if dist=='cq':
        nax=2
    else:
        nax=numcond
    f, axes = plt.subplots(1,nax,figsize=(nax*4,4),sharey=False)
    axes = np.array(axes).flatten()

    if c is None:
        c = sns.color_palette('hot', n_colors=numcond)
        
    for i, (cond, cdf) in enumerate(df.groupby(group)):
        
        cc=c[i]
        rts=cdf.rt.values
        rtvector = rts[rts<=filtr]

        if 'q' in dist:
            ax = axes[0]
            if cond==60:
                cond='<60'
            ax.plot(mq(rtvector, prob=prob), prob,  color=cc, label=cond)
            axes[0].set_title('Quantile-Probability/Cost F(x)')
            if 'c' in dist:
                sns.kdeplot(rtvector, color=cc, cumulative=True, 
                            label=str(cond), ax=axes[1], clip=xlim)
                axes[1].set_title('Estimated Cumulative')
                continue
                
        elif dist=='c':
            ax = axes[0]
            sns.kdeplot(rtvector, color=c[i], cumulative=True, label=str(cond), ax=ax)
            
        elif dist=='h':
            sns.distplot(rtvector, ax=axes[i], bins=bins, label=str(cond),
                         hist_kws={'color':cc, 'normed':False}, kde=False)
            ax = axes[i]
        elif dist=='k':
            sns.kdeplot(rtvector, ax=axes[i], color=c[i], label=str(cond), shade=True)
            ax=axes[i]
        else:
            print str(cond) + ': Mean RT=%.4f' % rtvector.mean()
            print """dist values: "k" (kdeplot), "h" (histogram), 
            "c" (cumulative), "qp" (qp plot), or "cq" (cumulative and qp)"""
            
        print str(cond) + " Mean RT: %.5f" % np.mean(rtvector)
        
    for ax in axes:
        ax.set_xlim(xlim)
        ax.legend(loc=2)
    
    plt.tight_layout()
    sns.despine()
    return f
```

```python
from __future__ import division

def plot_perc_rt(df, x1, x2, ax=None, filtr=.54502, res=2000, tagRT=True):
    x = np.linspace(x1, x2, res)
    go = df.query('response==1')
    y = [len(go[go.rt<=yy])/len(go.rt) for yy in x]
    ax.plot(x, y, lw=0, marker='o', markersize=2)
    ax.plot(filtr,len(go[go.rt<=filtr])/len(go.rt), lw=0, marker='o', color='r', markersize=9, label='RT cutoff for fitting')
    
    ax.legend(loc=2)
    
    muresp = df.response.mean()
    choice_cutoff = x[np.argmax(y>=muresp)]
    ax.vlines(choice_cutoff, ymin=0, ymax=1, color='k', alpha=.4, linestyle='--', label='choice cutoff')
    
    if tagRT:
        tag=filtr
    else:
        tag=choice_cutoff
        
    
    plt.setp(ax, ylabel='% Go trials left', xlabel='RT cutoff', xticks=[x1, tag, x2], xticklabels
             =[x1, str(tag)[:7], x2], xlim=[x1, x2])
    plt.legend(loc=0)
```
