```python
def plot_traces(DVg, DVs, sim_theta, plot_ss=True, tau=.0005, tb=.650, cg='Green', cr='Red'):
    
    nss=len(DVs)
    f,ax=plt.subplots(1,figsize=(8,5))
    tr=sim_theta['t']; a=sim_theta['a']; z=sim_theta['z']; ssd=sim_theta['ssd']

    for i, igo in enumerate(DVg[nss:]):        
        gtrace = igo[igo<=a]
        xg = tr+np.arange(len(gtrace))*.0005
        plt.plot(xg, gtrace, color=greens[2], linewidth=.5, alpha=.05)
    
    if 'ab' in sim_theta.keys():
        a=a+sim_theta['ab']
    
    for i, igo in enumerate(DVg[:nss]):
        gtrace = igo[igo<=a]
        strace = DVs[i][DVs[i]>=0]
        
        xg = tr+np.arange(len(gtrace))*.0005
        plt.plot(xg, gtrace, color=blues[4], linewidth=.5, alpha=.05)
        
        if plot_ss:
            #if i<len(DVs):
            xs = ssd+np.arange(len(strace))*.0005
            #return xs
            plt.plot(xs, strace, color=reds[1], alpha=.05, linewidth=.1)
    
    xlim=(tr-(.15*tr), tb+(.05*tr))
    ylim=(-.2, ax.get_ylim()[1])
    plt.setp(ax, xlim=xlim)
    ax.hlines(y=z, xmin=xlim[0], xmax=tb, linewidth=2, linestyle='--', color="#6C7A89")
    ax.hlines(y=a, xmin=ssd, xmax=tb, linewidth=2, linestyle='-', color="#e74c3c")
    ax.hlines(y=a-theta['ab'], xmin=xlim[0], xmax=ssd, linewidth=2, linestyle='-', color="k")
    ax.hlines(y=a-theta['ab'], xmin=ssd, xmax=tb, linewidth=2, linestyle='-', color='k', alpha=.3)
    ax.vlines(x=ssd, ymin=a-theta['ab'], ymax=a, linewidth=2, linestyle='-', color="#e74c3c")
    ax.hlines(y=0, xmin=xlim[0], xmax=tb, linewidth=2, linestyle='-', color="k")
    ax.hlines(y=-.05, xmin=xlim[0], xmax=tb, linewidth=2, linestyle='-', color="k", alpha=.0)
    sns.despine(top=False,bottom=False, right=True, left=False)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_xticks([])
    ax.set_yticks([])

    return ax
```
