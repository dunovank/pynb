```python
qsim=reload(qsim); ss=reload(ss)
pth=utils.find_path()

vfull_stats=pd.read_csv(pth+"CoAx/SS/HDDM/Proactive/vfull_sx/vfull_HiLo_550_SxStats.csv")
p=ft.stats_summary(vfull_stats)['mean'].to_dict()

vlist=[p['v(Lo)'], p['v(Hi)']]
gp={'a':p['a'], 'z':p['z'], 'Ter':p['t'], 'eta':p['sv'], 'st':0.000001, 'sz':0.000001}
sp={'mu_ss':-1, 'ssTer':0.0, 'ssTer_var':0.0}

ssvlist=-1*np.arange(1, 2.5, .5)
pGo_Lo=np.ones_like(ssvlist)*.2

vlist_lo=np.ones_like(ssvlist)*vlist[0]
vlist_hi=np.ones_like(ssvlist)*vlist[1]
conditions=["ssv"+str(v) for v in ssvlist]

LoSims=[]
for ssv in ssvlist:
    sp['mu_ss']=ssv
    sim_ssv_df=qsim.simConditions(gp, sp, vlist_lo, pGo_list=pGo_Lo, ssdlist=[.450], mfx=simfx.thal, 
                   ntrials=100, tb=.560, task='ssPro', conditions=conditions)
    
    LoSims.append(sim_ssv_df)

lo_df=pd.concat(LoSims)
```
