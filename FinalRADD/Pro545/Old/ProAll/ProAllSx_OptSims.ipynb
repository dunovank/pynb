```python
def make_hi_lo_col(df):
    
    dfl=df.query('pGo<.5')
    dfh=df.query('pGo>.5')
    
    dfl['HiLo']=['Lo']*len(dfl)
    dfh['HiLo']=['Hi']*len(dfh)
    
    return pd.concat([dfl, dfh])

def get_hilo_means(df):
    
    df.groupby(['HiLo', 'response']).mean().reset_index()
    
```
