```python
rebsl=pd.read_csv("CoAx/SS/DataFiles/ReBSL_Data.csv", index_col=0)
df=rebsl.copy()
sacc=df[df['trial_type']=='stop'].groupby('ssd').mean()['acc'].values
godf=df[(df['trial_type']=='go')&(df['acc']==1)&(df['rt']<.650)]
mrt=godf.rt.mean()
y=np.append(sacc, mrt)
```

```python
tparams=pd.read_csv("CoAx/SS/tbias/ReHDDM_tbias_params.csv")
tinits=tparams.loc[:, 'a':].mean()
tbsl=tinits.drop('t(pnl)').rename({'t(bsl)':'t'})
tpnl=tinits.drop('t(bsl)').rename({'t(pnl)':'t'})

vparams=pd.read_csv("CoAx/SS/vbias/ReHDDM_vbias_params.csv")
vinits=vparams.loc[:, 'a':'z'].mean()
vbsl=vinits.drop('v(pnl)').rename({'v(bsl)':'v'})
vpnl=vinits.drop('v(bsl)').rename({'v(pnl)':'v'})

old=pd.read_csv("CoAx/SS/ParamFiles/ReBSL_RaDD_Params.csv")
old_sx_ssvs=old['ssv']
ssv=-old_sx_ssvs.mean()

for params in [tbsl, tpnl, vbsl, vpnl]:
    params['ssv']=ssv
```

```python
chunk_list = lambda lst, siz: [lst[i:i+siz] for i in range(0, len(lst), siz)]
chunk_df = lambda lol, df: [df[df['idx'].isin(l)] for l in lol]
idxmean = lambda df: df.groupby('idx').rt.mean().values


sxlist=list(rebsl.idx.unique())
idchunks=chunk_list(sxlist, 12)
idchunks[-2].append(idchunks.pop(-1)[0])
dflist=chunk_df(idchunks, rebsl)
```

```python
tparams=pd.read_csv("CoAx/SS/tbias/ReHDDM_tbias_params.csv")
tinits=tparams.loc[:, 'a':].mean()
tbsl=tinits.drop('t(pnl)').rename({'t(bsl)':'t'})
tpnl=tinits.drop('t(bsl)').rename({'t(pnl)':'t'})

tbsl['ssv']=ssv
tpnl['ssv']=ssv

flatbsl=rebsl.copy()
flatbsl['idx']=['flat']*len(flatbsl)
```

```python
vbslx=vbsl[['a', 'v', 't', 'ssv', 'z']]
vbslx['ssv']=-.6
```

```python
all_proc_stats = allproc.map(fitre.fit_reactive_sx(inits=tbsl.to_dict(), depends=['t'], ntrials=1000, maxfun=5, depHyper=True, ftol=1.e-1, xtol=1.e-1), dflist[:])
```

```python
ssvlist=-np.arange(.2,3.0,.2)

yh=[]; reslist=[]; errlist=[]

for ssv in ssvlist:
    vbsl['ssv']=ssv
    
    df=fitre.simConditions(vbsl, visual=True, ntrials=500, hyper_delay=0.0)
    
    yh=list(df[df['trial_type']=='stop'].groupby('ssd').mean()['acc'][:])
    godf=df[(df['trial_type']=='go')&(df['acc']==1)&(df['rt']<.650)]
    yh.append(godf.rt.mean())
    
    yhat=np.array(yh)
    res=yhat-y 
    err=np.sum((yhat-y)**2)
    reslist.append(res)
    errlist.append(err)
```
