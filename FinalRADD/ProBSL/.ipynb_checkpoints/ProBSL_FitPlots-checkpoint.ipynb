```python
paramdf=pd.read_csv(utils.find_path()+"fullRaDD/ReBSL/kdfitted_sxparameters.csv", index_col='subj_idx')
rePar=pd.read_csv(utils.find_path()+"HDDM/Reactive/vbias_full/vBP_SxStats.csv")

hmPar=utils.makePivot(rePar, cols='param', rows='subj_idx', values='mean')
hmPar['a']=hmPar['a']*.1
hmPar['v(bsl)']=hmPar['v(bsl)']*.1
hmPar['v(pnl)']=hmPar['v(pnl)']*.1

inter=pd.Series(list(set(paramdf.index.values).intersection(set(hmPar.index.values))))
inter

redf=paramdf[paramdf.index.isin(inter)]
hmdf=hmPar[hmPar.index.isin(inter)]

#np.correlate()
```

```python
pcor_dict={}
f,axes=plt.subplots(2,2, figsize=(8, 8))

flatui = ["#9b59b6", "#3498db","#e74c3c", "#009B76", "#e74c3c",
          "#34495e", "#2ecc71","#95a5a6","#4168B7"]

params=['a', 'v', 'Ter', 'z']
i=0
for parameter in params:
    if parameter=='v':
        x=hmdf['v(bsl)']; y=redf[parameter]
        x=x.drop(54); y=y.drop(54)
    elif parameter=='Ter':
        x=hmdf['t']; y=redf[parameter]
    elif parameter=='z':
        x=hmdf['a']*hmdf['z']; y=redf[parameter]
        x=x.drop(38); y=y.drop(38)
    else:
        x=hmdf[parameter]; y=redf[parameter]
           
    pcor=np.corrcoef(x,y)[0][1]
    m,b=np.polyfit(x,y,1)
    
    corr_str="%s: r=%s" % (parameter, str(pcor)[:5])
    ax=axes.flatten()[i]
    ax.scatter(x, y, color=flatui[i])
    ax.plot(x, m*x+b,color=flatui[i], label=corr_str)
    ax.legend(loc=2, fontsize=14)
    pcor_dict[parameter]=pcor
    
    sns.despine()
    if i==0 or i==2:
        plt.setp(ax, ylabel='RaDD')
    if i==2 or i==3:
        plt.setp(ax, xlabel='HDDM')
    i+=1

#plot outliers
z_ax=axes.flatten()[3]
x=hmdf['a']*hmdf['z']; y=redf['z']
z_ax.scatter(x.loc[38], y.loc[38], color='k')

v_ax=axes.flatten()[1]
x=hmdf['v(bsl)']; y=redf['v']
v_ax.scatter(x.loc[54], y.loc[54], color='k')


plt.tight_layout()
f.savefig("ReBSL_HDDM_ParamsCorrelation.png", dpi=600)
```
