```python
from __future__ import division
from radd import ss, ssex, ssex_onsets, psy, thalamus
import seaborn as sns
import pandas as pd
import os
import collections
from itertools import cycle

"""
For the reactive condition, the stop signals occurred at 

75ms (early stop)
250ms (late stop)

For the proactive condition, the stopping probabilities were:

0%   (Go trials)
33% (low prob)
67% (high prob)
100% (All Stop)
"""

#globals
a=.37; z=.5*a; Ter=.347; eta=.14; st=.0001; sz=.0001; s2=.01; xpo=[12, 12.29]
mu_ss=-2.17; ssTer=.099; ssRe_TB=.653; ssPro_TB=.6; nt=3000; sTB=.00001; ssTer_var=.0001

#ssRe
#ssdlist=[.2, .25, .3, .35, .4]
#ssPro
#pGo_list=[0, .2, .4, .6, .8, 1]
pSSD=.450
#vlist_bsl=[0.073, 0.26, 0.4385, 0.601, 0.93, 1.09]
#vlist_pnl=[0.05, 0.23, 0.425, 0.601, 0.93, 1.09]
```

```python
ssex=reload(ssex)
ss=reload(ss)

pGo=[0, .33, .67, 1]
vlist=[0.073, 0.38, 0.71, 1.09]

df_v=[]

for i,v in enumerate(vlist):
    gp={'a':a, 'z':z, 'v':v, 'Ter':Ter, 'eta':eta, 'st':st, 'sz':sz, 's2':s2}
    sp={'mu_ss':-2.17, 'pGo':pGo[i], 'ssd':.450, 'ssTer':.099, 'ssTer_var':.0001}
    
    dfout=ss.set_model(gParams=gp, sParams=sp, ntrials=nt, timebound=ssRe_TB, stb=sTB, t_exp=True, 
                       analyze=True, exp_scale=xpo, visual=False, task='ssProBSL', 
                       predictBOLD=True, test_predictBOLD=False)
    
    df_v.append(dfout)
```

```python
conditions=['pg0', 'pg33', 'pg67', 'pg100']
p_dict=dict()

for i, dfout in enumerate(df_v):
    
    conds_dict=dict()
    
    for df in dfout.keys():
        dfout[df].dropna(inplace=True)
        dfout[df].index=np.arange(len(dfout[df]))
        
        minlist=[]; maxlist=[]

        for trace in dfout[df][:]:
            csum_min=pd.Series(trace).cumsum().min()
            csum_max=pd.Series(trace).cumsum().max()
            minlist.append(csum_min)
            maxlist.append(csum_max)
        
        conds_dict[df]=(np.mean(minlist), np.mean(maxlist))
        
    p_dict[conditions[i]]=conds_dict
```

```python
#split by whether a stop or a go trial first then show trends for pgo0-67 in stop trials and pgo33-100 in go trials

for d in p_dict.keys():
    for k in p_dict[d].keys():
        print "%s ms; %s: %s\n" % (d, k, str(p_dict[d][k][1]))
```
