```python
from __future__ import division
from radd.simfx import sim_radd, sim_ddm
from radd import ss, simfx, psy
import pandas as pd
import os
import scipy.stats as stats
import seaborn as sns

#globals
a=.37; z=.5*a; Ter=.347; eta=.14; st=.0001; sz=.0001; s2=.01; xpo=[12, 12.29]
mu_ss=-2.17; ssTer=.099; ssRe_TB=.653; ssPro_TB=.6; nt=5000; sTB=.00001; ssTer_var=.0001
labels=['P(Go)=80', 'P(Go)=60', 'P(Go)=40', 'P(Go)=20', 'P(Go)=0']
```

```python
### Standard Model Simulations ###

ss=reload(ss); simfx=reload(simfx); psy=reload(psy)

Sdfs=[]

ssdlist=[.200, .250, .300, .350, .400, .450]

gp={'a':a, 'z':z, 'v':0.611, 'Ter':Ter, 'eta':eta, 'st':st, 'sz':sz, 's2':s2}

for i, ssd in enumerate(ssdlist):

    sp={'mu_ss':mu_ss, 'pGo':.5, 'ssd':ssd, 'ssTer':ssTer, 'ssTer_var':ssTer_var}
    
    #simulate and predict bold using modelfx.sustained_integrator()
    Stndrd_df=ss.set_model(gParams=gp, sParams=sp, mfx=sim_ddm, ntrials=nt, timebound=ssRe_TB, t_exp=True, 
                       exp_scale=xpo, visual=False, task='ssReBSL', predictBOLD=False, return_dataframe=True)
    
    Sdfs.append(Stndrd_df)
```

```python
### Calc 95%CI Upper Boundary for Standard ###

for i, d in enumerate(Sdfs):
    d['SSD']=[ssdlist[i]]*len(d)

bigS=pd.concat(Sdfs)

rt_vectorsS=[]
ci_listS=[]

for cond, df in bigS.groupby('SSD'):
    
    data=df.query("trial_type == 'stop' and acc == 0")
    
    Smu = data.rt.mean()
    Ssigma = data.rt.std()
    
    rt_vectorsS.append(data['rt'])
    
    CI=stats.norm.interval(.95, loc=Smu, scale=Ssigma)
    ci_listS.append(CI[1])
    
### Calc Delay Bias for Standard ###
Sbias=[x-ssdlist[i] for i, x in enumerate(ci_listS)]
```
