## Computational Methods for Data Analysis - Lecture 2

```python
import scipy as sp
from numpy.fft import *
```

```python
T = 30
n = 512

t = linspace(-T/2., T/2., n, endpoint=False)
```

```python
k = (2*pi/T)*r_[linspace(0, n/2-1, n/2), linspace(-n/2,-1,n/2)]
```

```python
# Signal as function of time
sech = lambda x: 1./(np.cosh(x))
u = sech(t)
ut = fft(u)
```

```python
subplot(211)
plot(t,u)
subplot(212)
plot(fftshift(k), abs(fftshift(ut))/max(abs(fftshift(ut))))
xlim(-15,15)
```

Now, add noise to every frequency component (white noise)

```python
noise = 10
utn = ut + noise*(randn(n) + 1j*randn(n))
# Transform back to time domain
un = ifft(utn)
```

```python
subplot(211)
plot(t,abs(un),t,u,'r--')
subplot(212)
plot(fftshift(k), abs(fftshift(utn))/max(abs(fftshift(utn))),fftshift(k), abs(fftshift(ut))/max(abs(fftshift(ut))),'r--')
xlim(-15,15)
```

### Denoising

Can we retrieve the signal from the above?

```python
filter_ = exp(-k**2)

plot(fftshift(k), abs(fftshift(utn))/max(abs(fftshift(utn))),fftshift(k), fftshift(filter_), 'r--')
```

Multiply the noisy signal by the filter

```python
utnf = filter_ * utn

plot(fftshift(k), abs(fftshift(utn))/max(abs(fftshift(utn))))
plot(fftshift(k), fftshift(filter_), 'r--')
plot(fftshift(k), abs(fftshift(utnf))/max(abs(fftshift(utnf))), 'k--')
xlim(-5,5)
```

```python
unf = ifft(utnf)

plot(t,abs(un),alpha=0.2)
plot(t,abs(unf),'k--',t,[0.5]*len(t),'r')
```

Shift the filter.

```python
filter_ = exp(-(k+15)**2)

utnf = filter_ * utn

unf = ifft(utnf)

subplot(211)
plot(t,abs(un),alpha=0.2)
plot(t,abs(unf),'k--',t,[0.5]*len(t),'r')
subplot(212)
plot(fftshift(k), abs(fftshift(utn))/max(abs(fftshift(utn))))
plot(fftshift(k), fftshift(filter_), 'r--')
plot(fftshift(k), abs(fftshift(utnf))/max(abs(fftshift(utnf))), 'k--')
xlim(-20,20)
```
