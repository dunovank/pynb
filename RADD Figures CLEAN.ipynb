```python
from radd import utils, simfx, ss, psy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
import seaborn as sns
```

```python
def find_path():

    if os.path.isdir("/Users/kyle"):
        pth="/Users/kyle/Dropbox/"

    elif os.path.isdir("/home/kyle"):
        pth="/home/kyle/Dropbox/"

    return pth

def simConditions(theta, mfx=simfx.sim_radd, ntrials=1500, 
                  return_all=True, tb=.650, task='ssRe'):
    theta['pGo']=.5
    simdf_list=[]; ssdlist=np.arange(.2,.4,.05)
    for ssd in ssdlist:
        
        theta['ssd']=ssd
        out=ss.set_model(theta, mfx=mfx, ntrials=ntrials, visual=True, 
                         intervar=True, return_all=True, timebound=tb) 

        simdf_list.append(out)
    return pd.concat(simdf_list)

def get_reactive_stops(df):
    avg=df.groupby('ssdlist').mean()
    sem=df.groupby('ssdlist').sem()

    yd=avg['ydata'][:-1]
    yh=avg['yhat'][:-1]
    yderr=sem['ydata'][:-1]*1.96
    yherr=sem['yhat'][:-1]*1.96
    
    return yd, yh, yderr, yherr


def get_reactive_rts(df):
    
    avg=df.groupby('ssdlist').mean()
    sems=df.groupby('ssdlist').sem()
    
    rtd=avg['ydata'][-1]/10
    rth=avg['yhat'][-1]/10
    
    rtd_err=sems['ydata'][-1]/10*1.96
    rth_err=sems['yhat'][-1]/10*1.96
    
    ys=rtd
    
    return rtd, rth, rtd_err, rth_err, ys

def makePivot(df, cols='ssd', index=None, values='stop_acc', func=np.mean):
    pvot=pd.pivot_table(df, index=index, columns=cols, values=values, aggfunc=func)
    return pvot

def pgo(df):
    return makePivot(df, cols='pGo', index='idx', values='response')

def pstop(df):
    return 1-makePivot(df, cols='pGo', index='idx', values='response')

def pro_rt(df):
    godf=df[df['choice']=='go']
    return makePivot(godf, cols='pGo', index='idx', values='rt')

def trunc_df(df):
    godf=df.query("choice=='go' and rt<=.545")
    stopdf=df.query("choice=='stop' and rt>.545")
    return pd.concat([godf, stopdf])

def re_rt(df):
    corr=df[(df['choice']=='go')&(df['acc']==1)]
    return makePivot(corr, cols='ssd', index='idx', values='rt')

def re_acc(df):
    stops=df[df['choice']=='stop']
    return makePivot(stops, index='idx', cols='ssd', values='acc')


pth=find_path()
raddpath=pth
datapath=pth+"CoAx/SS/DataFiles/"
fitspath=pth+"radd_SFN14/"
styl=utils.style_params()
sns.set_style('ticks', rc=styl['style'])
sns.set_context('paper', rc=styl['context'])
sns.set_palette(styl['colors'])
```

```python
import numpy as np
import scipy as sp
import scipy.stats

def get_ci(data, confidence=0.95):
    a = 1.0*np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * sp.stats.t._ppf((1+confidence)/2., n-1)
    return h
```
