```python
def get_slopes(df, start, direction='pos'):
    slopes=[]
    for i, (sx, sxvec) in enumerate(df.groupby('sx')):
        
        xhigh=np.where(sxvec.values==sxvec[start].values)[1][0]
        yhigh=sxvec[start].values[0]
        
        if direction=='pos':
            xlow=np.where(sxvec.values==sxvec[start].values)[1][0]
            ylow=sxvec[start].values[0]
            xhigh=np.where(sxvec.values==np.max(sxvec.ix[:,start:].values[0][1:]))[1][0]
            yhigh=np.max(sxvec.ix[:,start:].values)

        elif direction=='neg':
            xlow=np.where(sxvec.values==np.min(sxvec.ix[:,start:].values[0][1:]))[1][0]
            ylow=np.min(sxvec.ix[:,start:].values)
            xhigh=np.where(sxvec.values==sxvec[start].values)[1][0]
            yhigh=sxvec[start].values[0]
        
        else: 
            print "direction must be string, 'pos' or 'neg'"
            return []
        
        slopes.append((yhigh-ylow)/(xhigh-xlow))
    return np.array(slopes)    


def plot_regression(x, y, robust=True, color='Navy', statfunc=pearsonr, ax=None):
    if ax is None:
        f, ax = plt.subplots(1)
    xl=(x.min()-(.05*x.min()), x.max()+(.05*x.max()))
    yl=(y.min()-(.05*y.min()), y.max()+(.05*y.max()))
    sns.regplot(x, y, robust=robust, line_kws={"color":color}, 
                scatter_kws={'color':color, 'alpha':.8}, ax=ax)
    # annotate
    l=ax.lines; 
    rho=statfunc(x, y)
    cis=boot.ci((x, y), statfunction=statfunc, alpha=.05)
    ax.legend(l, ['rho=%.3f, p=%.3f\n95CI=[%.3f,%.3f]' % (rho[0], rho[1], cis[0,0], cis[1,0])], fontsize=16, loc=0)
    sns.despine(); ax.set_xlim(xl); ax.set_ylim(yl); plt.tight_layout()
```
