```python
import hddm
import pandas as pd
import kabuki
```

```python
data=pd.read_csv("/home/kyle/Dropbox/PFH/iPFH/HDDM/Final/allsx_ewma.csv")
models = []
for i in range(1,3):
    dbm = hddm.HDDM(data, depends_on={'v':['stim', 'cue']},  bias=False, informative=False, include=['v', 't', 'a', 'sv', 'sz', 'st'])
    dbm.sample(5000, burn=1000, dbname="dbm_traces"+str(i)+".db", db='pickle')
    dbm.print_stats("stats"+str(i)+".txt")
    models.append(dbm)
allmodels = kabuki.utils.concat_models(models)
allmodels.plot_posteriors(save=True)
allmodels.print_stats("combined_model_stats.txt")
```
