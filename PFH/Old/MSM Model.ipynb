```python
import hddm
import pandas as pd
import kabuki
```

```python
#start commands
data=pd.read_csv("/home/kyle/Dropbox/PFH/iPFH/HDDM/Final/allsx_ewma.csv")
models = []

for i in range(1,3):
    msm = hddm.HDDM(data, depends_on={'v':['stim', 'cue'], 'z':'cue'},  bias=True, informative=False, include=['v','z','t','a','sv', 'sz', 'st'])
    msm.sample(5000, burn=1000, dbname="msm_traces"+str(i)+".db", db='pickle')
    msm.print_stats("stats"+str(i)+".txt")
    models.append(msm)

allmodels = kabuki.utils.concat_models(models)
allmodels.plot_posteriors(save=True)
allmodels.print_stats("combined_model_stats.txt")
```
