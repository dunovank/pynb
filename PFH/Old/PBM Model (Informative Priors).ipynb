```python
import hddm
import pandas as pd
import kabuki
```

```python
#PBM
%cd /home/kyle/Dropbox/PFH/iPFH/HDDM/Final/pbm_info
data=pd.read_csv("/home/kyle/Dropbox/PFH/iPFH/HDDM/Final/allsx_ewma.csv")
params=['v', 'z', 't', 'a', 'sv', 'sz', 'st']
```

```python
models = []
for i in range(3):
    
    pbm = hddm.HDDM(data, depends_on={'v':'stim', 'z':'cue'}, bias=True, include=params)
    pbm.sample(5000, burn=1000, dbname="pbm_traces"+str(i)+".db", db='pickle')
    pbm.print_stats("stats"+str(i)+".txt")
    models.append(pbm)
    
allmodels = kabuki.utils.concat_models(models)
allmodels.plot_posteriors(save=True)
allmodels.print_stats("combined_model_stats.txt")
allmodels.save('pbm')
```

```python
gr=kabuki.analyze.gelman_rubin(models)
grdf=pd.DataFrame.from_dict(gr, orient='index')
grdf.rename(columns={0:'GelmanRubin'}, inplace=True)
grdf.to_csv('gelman_rubin.csv')
```
