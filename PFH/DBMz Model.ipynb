```python
#start commands
data=pd.read_csv("/home/kyle/Dropbox/PFH/iPFH/HDDM/Final/allsx_ewma.csv")
models = []
for i in range(1,3):
    dbmz = hddm.HDDM(data, depends_on={'v':['stim', 'cue']},  bias=True, informative=False, include=['v', 'z', 't', 'a', 'sv', 'sz', 'st'])
    dbmz.sample(5000, burn=1000, dbname="dbmz_traces"+str(i)+".db", db='pickle')
    dbmz.print_stats("stats"+str(i)+".txt")
    models.append(dbmz)
allmodels = kabuki.utils.concat_models(models)
allmodels.plot_posteriors(save=True)
allmodels.print_stats("combined_model_stats.txt")
kabuki.analyze.gelman_rubin(models)
```
