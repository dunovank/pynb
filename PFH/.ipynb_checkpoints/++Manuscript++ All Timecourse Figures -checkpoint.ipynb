# WIN Cues

```python
labels=['Face Cue', 'Neutral Cue', 'House Cue']  
f,axes=plt.subplots(1,2, figsize=(10,5))

for ii, win in enumerate([wffa, wppa]):
    renamecols={str(i):int(i) for i in range(1,13)}
    win.rename(columns=renamecols, axis=1, inplace=True)
    ax=f.axes[ii]
    
    dfcue=win[win['cond'].isin(['hc', 'nc', 'fc'])]
    dfcue.replace({'cond':{'hc':3, 'nc':2, 'fc':1}}, inplace=True)
    cue=dfcue.groupby('cond').mean().loc[:, 1:12]
    cue_sem=dfcue.groupby('cond').sem().loc[:, 1:12]
    #cue_sem.drop('sx', axis=1, inplace=True)
    
    cmeans=[cue.loc[i, :].values for i in cue.index.unique()]
    cerrors=[cue_sem.loc[i, :].values for i in cue_sem.index.unique()]
    x=np.arange(len(cue.columns))
    
    if ii==0:
        colors=["#0000B8",'#005CE6',"#89a0ff"]
    else:
        colors=["#990000" ,'#FF1919', "#ea8989"]
        cmeans=cmeans[::-1]
        labels=labels[::-1]

    for i, y in enumerate(cmeans):
        ax.plot(x, y, color=colors[i], linewidth=4.5)
        ax.fill_between(x, y-cerrors[i], y+cerrors[i], color=colors[i], alpha=.2)
        ax.hlines(y=0.0, xmin=0, xmax=12, color='#575757', linestyle='--')
    
    if ii==0:
        plt.setp(ax, ylim=(-.1, .4), yticks=[-.1, .0, .4], yticklabels=[-.1, .0, .4], 
             xlim=(0,11),xticks=np.arange(0,12, 1), xticklabels=np.arange(0,24,2))
        ax.set_title('Face ROIs', fontsize=32)
        ax.set_ylabel("BOLD (PSC)", fontsize=25)
    
    else:
        ax.set_title('House ROIs', fontsize=32)
        plt.setp(ax, ylim=(-.1, .4), yticks=[-.1,.0,.4], yticklabels=[-.1,.0,.4], 
                 xlim=(0,11),xticks=np.arange(12), xticklabels=np.arange(0,24,2))

    plt.setp(ax.get_yticklabels(), fontsize=20)
    ax.set_xlabel("Time (sec)", fontsize=25)
    plt.setp(ax.get_xticklabels(), fontsize=18)
    f.subplots_adjust(wspace=.08)
    plt.tight_layout()
    sns.despine()
#f.savefig('/home/kyle/Desktop/Cue_FWIN_HWIN.png', dpi=600) 
```
